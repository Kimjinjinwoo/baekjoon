```python
## O(N^2)으로 생각해 봐도 될 듯..?
## N이랑 M 최댓값이 1000이라 가지고
## 가능한 것들 중에서 끝이 제일 짧은거로 하면 될 듯~ 
## 이분 매칭으로도 풀어보자!
import copy
def sol(N, M, applications):
    ans = 0
    applications.sort()
    candi = []
    for x in range(1,N+1):
        candi = filtering(x, candi)
        cnadi = insert(x, candi, applications)
        if candi:
            candi.sort(key = lambda x: x[1])
            candi.pop(0)
            ans += 1
    print(ans)
def filtering(x, candi):
    next_candi = []
    for appli in candi:
        if appli[0] <= x <= appli[1]:
            next_candi.append(appli)
    return next_candi

def insert(x, candi, applications):
    while applications:
        if applications[0][0] == x:
            candi.append(applications.pop(0))
        else:
            break
    return candi

sol(N, M, copy.deepcopy(applications))
```
