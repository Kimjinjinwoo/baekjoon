```python
def get_input():
    N, C = map(int, input().split())
    nums = list(map(int, input().split()))

    return N, C, nums

def split_list(N, nums):
    N1 = N // 2; N2 = N - N//2
    nums1 = nums[:N1]
    nums2 = nums[N1:]

    return N1, nums1, N2, nums2


if __name__ == "__main__":
    N, C, nums = get_input()

    N1, nums1, N2, nums2 = split_list(N, nums)


```
