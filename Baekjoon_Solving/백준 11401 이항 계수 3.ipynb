```python
def factorial(N, MOD = 1):
    res = 1
    for ii in range(1, N+1):
        res *= ii
        res %= MOD
    return res

def n_square(N, number, MOD = 1):
    if N == 1:
        return number

    sqrt_number = n_square(N // 2, number, MOD = MOD)
    result = sqrt_number * sqrt_number
    if N % 2 == 1: ## 홀수라면
        result *= number
    return result % MOD

if __name__ == "__main__":
    N, K = map(int, input().split())
    MOD = 1000000007

    N_fac = factorial(N, MOD = MOD)

    K_fac = factorial(K, MOD = MOD)
    N_minus_K_fac = factorial(N-K, MOD = MOD)

    inverse = n_square(MOD-2, K_fac * N_minus_K_fac,  MOD = MOD)

    print((inverse * N_fac) % MOD)
```
