```python
import sys

sys.setrecursionlimit(10 ** 9)

M , N , K = map( int , input().split() )

space= [[False] * N for _ in range(M)]

#뒤집어서 생각하자! 즉 (0,0)이 좌상단
def squre(x_1,y_1,x_2,y_2):
    for i in range(y_1 , y_2):
        for j in range(x_1,x_2):
            space[i][j] = True


#직사각형
for _ in range(K):
    x_1,y_1,x_2,y_2 = map(int , input().split())
    squre(x_1,y_1,x_2,y_2)



#dfs
def dfs(i,j):
    global area
    if not space[i][j]:
        space[i][j] = True
        area += 1
        i_co = [i,i,i+1,i-1]
        j_co = [j-1,j+1,j,j]

        for ii,jj in zip(i_co , j_co):
            if 0<=ii<M and 0<=jj<N:
                dfs(ii,jj)

many = 0
areaList =[]

for i in range(M):
    for j in range(N):
        if not space[i][j]:
            many+=1
            area = 0
            dfs(i,j)
            areaList.append(area)

print(many)
areaList.sort()
for area in areaList:
    print(area, end = ' ')
```
