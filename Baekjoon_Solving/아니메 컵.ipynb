```python
## 주문은 토기 입니까?
from collections import deque

def sol(N, M, arriveTimes):
    lastTime = max(arriveTimes)
    timeTable = [0] * (lastTime+1)

    for arriveTime in arriveTimes:
        if timeTable[arriveTime] == 0:
            timeTable[arriveTime] = 2
        else:
            print('fail')
            return
    needServeList = deque()
    for ii in range(lastTime, 0, -1):
        if timeTable[ii] == 2:
            needServeList.append(ii)
        elif timeTable[ii] == 0:
            if needServeList:
                needServe = needServeList.popleft()
                if needServe - M <= ii:                
                    timeTable[ii] = 1 #토기에 담기
                else:
                    print('fail')
                    return

    #print(timeTable)
    if needServeList:
        print('fail')
        return

    togiNum = 0
    for ii in range( lastTime + 1):
        if timeTable[ii] == 0:
            togiNum += 1
        elif timeTable[ii] == 1:
            if togiNum > 0:
                togiNum -= 1
            else:
                print('fail')
                return
    print('success')


N, M = map(int, input().split())
arriveTimes = list(map(int, input().split()))
sol(N, M, arriveTimes)
```
