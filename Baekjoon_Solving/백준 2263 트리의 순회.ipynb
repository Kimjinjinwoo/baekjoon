```python
## 스택이나 큐로 해야겠다.
import sys
from collections import deque

n = int(input())
in_order = list(map(int, input().split()))
post_order = list(map(int, input().split()))

graph = [[0, 0] for _ in range(n+1)]
ans = [0] * (n+1)
in_order_pos_dict = {}
for ii in range(n):
    in_order_pos_dict[in_order[ii]] = ii

def get_node_by_queue(in_order, post_order):
    queue = deque()
    node = (1, 0, n-1, 0, n-1) ## 0, len-1, 0, len-1
    queue.append(node)

    while(queue):
        pos, in_start, in_end, post_start, post_end = queue.pop() ## in_start, in_end, post_start, post_end
        #print(in_start, in_end, post_start, post_end)
        x = post_order[post_end] # post_end
        ans[pos] = x
        if in_start >= in_end:
            continue

        ii = in_order_pos_dict[x]
        cnt = ii - in_start

        if (ii+1 <= in_end) and (pos*2+1 <= n):
            if (ii+1 == in_end):
                ans[pos*2 + 1] = post_order[post_end-1]
            else:
                queue.append( (pos*2 + 1, ii+1, in_end, post_start + cnt, post_end-1) )

        if (in_start <= ii - 1) and (pos*2 <= n):
            if (in_start == ii-1):
                ans[pos*2] = post_order[post_start + cnt - 1]
            else:
                queue.append( (pos*2, in_start, ii - 1, post_start, post_start + cnt - 1) )

def main(n, in_order, post_order):
    get_node_by_queue(in_order, post_order)
    print(' '.join(map(str,ans[1:])))
    return

main(n, in_order, post_order)
```
