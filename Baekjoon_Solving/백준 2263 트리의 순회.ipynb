```python
## 스택이나 큐로 해야겠다.
import deque

n = int(input())
in_order = list(map(int, input().split()))
post_order = list(map(int, input().split()))

graph = [[0, 0] for _ in range(n+1)]

def get_node(in_order, post_order):
    if not post_order: ## post_order이 비어 있으면
        return 0
    x = post_order[ -1 ]

    ii = 0
    for ii, node in enumerate(in_order):
        if node == x:
            x_idx = ii
            break;
    
    left_in_order   = in_order[:ii]
    left_post_order = post_order[:ii]
    graph[x][0] = get_node(left_in_order, left_post_order)

    right_in_order  = in_order[ii+1:]
    right_post_order= post_order[ii:-1]
    graph[x][1] = get_node(right_in_order, right_post_order)

    return x
    
def get_node_by_queue(in_order, post_order):
    x = post_order[ -1 ]
    queue = deque()
    node = (x, in_order, post_order)
    queue.append(node)

    while(queue):
        x, in_order, post_order = queue.pop()

        left_in_order   = in_order[:ii]
        left_post_order = post_order[:ii]

        right_in_order  = in_order[ii+1:]
        right_post_order= post_order[ii:-1]
    

def pre_order(idx):
    if idx == 0:
        return
    print(idx, end = ' ')
    pre_order(graph[idx][0])
    pre_order(graph[idx][1])

def main(n, in_order, post_order):
    get_node(in_order, post_order)
    pre_order(post_order[-1])
    return

main(n, in_order, post_order)
```
