```python
## 스택이나 큐로 해야겠다.
## index로,,,,
from collections import deque

n = int(input())
in_order = list(map(int, input().split()))
post_order = list(map(int, input().split()))

graph = [[0, 0] for _ in range(n+1)]
    
def get_node_by_queue(in_start, in_end, post_start, post_end):
    queue = deque()
    node = (in_start, in_end, post_start, post_end)
    queue.append(node)

    while(queue):
        in_start, in_end, post_start, post_end = queue.popleft() 
        if not post_order:
            break
        x = post_order[post_end]

        ii = 0
        for ii, node_idx in enumerate( range(in_start, in_end+1) ):
            if in_order[node_idx] == x:
                x_idx = node_idx
                break;e

        if in_start != x_idx:
            graph[x][0] = post_order[x_idx-1]
            queue.append( (in_start, x_idx-1, post_start, x_idx-1) )

        if x_idx+1 != in_end:
            graph[x][1] = post_order[post_end-2]
            queue.append( (x_idx+1, in_end - 1, x_idx, post_end - 2) )

def pre_order(idx):
    if idx == 0:
        return
    print(idx, end = ' ')
    pre_order(graph[idx][0])
    pre_order(graph[idx][1])

def main(n, in_order, post_order):
    get_node_by_queue(0, n-1, 0, n-1)
    pre_order(post_order[-1])
    return

main(n, in_order, post_order)
```
