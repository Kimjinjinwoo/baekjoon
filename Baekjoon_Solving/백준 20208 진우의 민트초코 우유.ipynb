```python
N , M ,H = map(int , input().split())

milk = []
many = []

for i in range(N):
    row = list(map (int ,  input().split()) )
    for j , x in enumerate(row):
        if x == 2:
            milk.append([i,j])
        if x == 1:
            home = [i,j]

not_visited = milk[:]

def dist(a , b):
    d = abs(a[0]-b[0]) + abs(a[1]-b[1])
    return d

def find(now , hp , have , not_visited):
    hps = []
    locas = []
    idxs = []
    n = len(not_visited)


    for i in range(n):
        loca = not_visited[i]
        d = dist(now , loca)
        if d and (hp - d) >= 0:
            locas.append(loca)
            hps.append( hp - d + H)
            idxs.append(i)
    #한 블럭 더~
    
    for l, h,idx in zip(locas , hps, idxs):
        #여기 코드 개 구림 -> 인덱스로 하등가 해야할 듯
        no_visited = not_visited[:]
        no_visited.pop(idx)
        find(l , h , have+1 , no_visited)

    #go_home
    d_h = dist(now , home)
    if d_h and (hp - d_h) >= 0:
        many.append(have)
    
find(home , M , 0 , not_visited)

if many:
    print(max(many))
else:
    print(0)
```
