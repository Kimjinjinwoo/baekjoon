```python
import sys
sys.setrecursionlimit(10** 6)
input = sys.stdin.readline

def sol(N, links):
    tree = [[] for _ in range(N+1)]
    node_info = [[0,0]] * (N+1)
    for X, y, w  in links:
        tree[X].append( (y, w) )
    
    node_info = node_info_dfs(1, tree, node_info)

    maxi = 0
    for i in range(1,N+1):
        node_info[i].sort()
        if node_info[i][-1] + node_info[i][-2] > maxi:
            maxi = node_info[i][-1] + node_info[i][-2]
    print(maxi)

def node_info_dfs(now, tree, node_info):
    if not tree[now]: ## 자식 노드가 비어 있다면
        node_info[now] = [0,0]
    else: ##자식 노드가 있으면
        node = [0] * max(2, len(tree[now]))
        for i, son in enumerate(tree[now]):
            y, w = son
            node_info = node_info_dfs(y, tree, node_info)
            node[i] = max(node_info[y]) + w
        node_info[now] = node
    return node_info

N = int(input())
links = [list(map(int,input().split())) for _ in range(N-1)]
sol(N, links)
```
