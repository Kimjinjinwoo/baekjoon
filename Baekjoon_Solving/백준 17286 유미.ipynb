```python
import sys

def sol(yumi, persons):
    orders = [ [0,1,2], [0,2,1], [1,0,2], [1,2,0], [2,1,0], [2,0,1]]
    best = sys.maxsize
    for order in orders:
        now_persons = [persons[order[0]], persons[order[1]], persons[order[2]]]
        best = min(best, calc_straight(yumi, now_persons ))
    print(int(best))              

def calc_dist(A,B):
    Ax, Ay = A
    Bx, By = B
    dist = ( (Ax - Bx)**2 + (Ay - By)**2 ) ** (1/2)
    return dist


def calc_straight(yumi, persons):
    total = 0
    now = yumi
    for person in persons:
        total += calc_dist(now, person)
        now = person
    return total

yumi = list(map(int, map(int,input().split())))
persons = [list(map(int, input().split())) for _ in range(3)]
sol(yumi, persons)
```
