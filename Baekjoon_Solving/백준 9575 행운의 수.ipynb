```python
T = int(input())
counts = []

for _ in range(T):
    lucky = []
    count = 0
    lens = []
    nums = []
    for _ in range(3):
        lens.append(int(input()))
        nums.append(list( map( int , input().split())))
    nums_1 , nums_2 , nums_3 = nums
    nums_1 = list(set(nums_1))
    nums_2 = list(set(nums_2))
    nums_3 = list(set(nums_3))

    for num_1 in nums_1:
        for num_2 in nums_2:
            for num_3 in nums_3:
                sum_num = num_1 + num_2 + num_3
                sum_num = list(str(sum_num))
                sum_num = list(set(sum_num))
                sum_num.sort()
                if sum_num == ['5'] or sum_num == ['8'] or sum_num == ['5','8']:
                    lucky.append(num_1 + num_2 + num_3)
    lucky = list( set ( lucky ) )
    count = len(lucky)
    counts.append(count)

for ans in counts:
    print(ans)
```
