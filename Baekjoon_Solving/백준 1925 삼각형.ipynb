```python
def sol(A,B,C):
    angles = [calc_angle(A,B,C), calc_angle(B,A,C), calc_angle(C,A,C) ]
    ans = sum(angles)
    if ans == 3:
        print('JikkakTriangle')
    if ans == 2:
        print('')
    if ans == 1:
        print('')

def calc_angle(A,B,C):
    ## 0으로 옮기기.
    Ax,Ay = A
    Bx,By = B
    Cx,Cy = C
    Bx, By = Bx-Ax, By-Ay
    Cx, Cy = Cx-Ax, Cy-Ay
    cos = dot_product([Bx, By], [Cx, Cy]) / (calc_length(Bx,By) * calc_length(Cx, Cy))
    if cos > 0:
        return -1   # 예각
    elif cos == 0:
        return 0    #직각
    else:
        return 1    #둔각

def calc_length(x, y):
    return (x**2 + y**2)**(1/2)

def dot_product(X,Y):
    return X[0] * Y[0] + X[1] * Y[1]
```
