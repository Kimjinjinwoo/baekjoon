```python


L1 = list(map(int , input().split()))
L2 = list(map(int , input().split()))


def L1_function(x):
    x1 , y1 , x2 , y2 = L1
    m = (y1-y2)/(x1-x2)
    y = m * (x - x1) + y1
    return y


def L2_function(x):
    x1 , y1 , x2 , y2 = L2
    m = (y1-y2)/(x1-x2)
    y = m * (x - x1) + y1
    return y


def vali_not_verti(left_ , right_):
    v1 = L1_function(left_) - L2_function(left_) 
    v2 = L1_function(right_) - L2_function(right_) 
    if left_ > right_:
        print(0)
    elif v1 * v2 <= 0:
        print(1)
    else:
        print(0)


def vali_both_verti(left_ , right_):
    if L1[0] == L2[0]:
        if L1[1] <= L2[1] <= L1[3] or L1[1] <= L2[3] <= L1[3]:
            print(1)
            return
    print(0)


def vali_one_verti(verti , not_verti):
    global L1 ; global L2
    L1 = verti ; L2 = not_verti
    
    if L2[0] <= L1[0] <= L2[2]:
        if L1[1] <= L2_function(L1[0]) <= L1[3]:
            print(1)
            return
    print(0)
    return


def modify():
    global L1; global L2
    if L1[0] > L1[2]:
        L1 = [L1[2] , L1[3] , L1[0] , L1[1]]
    elif L1[0] == L1[2]:
        if L1[1] > L1[3]:
            L1 = [L1[2] , L1[3] , L1[0] , L1[1]]
    if L2[0] > L2[2]:
        L2 = [L2[2] , L2[3] , L2[0] , L2[1]]
    elif L2[0] == L2[2]:
        if L2[1] > L2[3]:
            L2 = [L2[2] , L2[3] , L2[0] , L2[1]]    


def vali(left_ , right_):
    if L1[0] == L1[2]:
        if L2[0] == L2[2]:
            vali_both_verti(left_ , right_)
        else:
            vali_one_verti(L1 , L2)
    elif L2[0] == L2[2]:
        vali_one_verti(L2 , L1)
    else:   # 둘 다 다름 
        vali_not_verti(left_ , right_)


modify()
left_   =   max(min(L1[0] , L1[2]) , min(L2[0] , L2[2]))
right_  =   min(max(L1[0] , L1[2]) , max(L2[0] , L2[2]))

vali(left_ , right_)
```
