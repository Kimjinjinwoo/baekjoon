```python
def get_NoPage(N, M, papers):
    paperList = [1] + [0] * (N)
    NoPage = []
    for paper in papers:
        paperList[paper] = 1
    for i,exist in enumerate(paperList):
        if not exist:
            NoPage.append(i)
    return NoPage

def sol(N, M, papers):
    NoPage = get_NoPage(N,M, papers)
    #print(NoPage)
    ink = 0
    if NoPage:
        start =  NoPage.pop(0)
        ink += 7
        for nopage in NoPage:
            if nopage - start >= 4:
                ink += 7
            else:
                ink += (nopage - start) * 2    
            start = nopage
    print(ink)

N,M = map(int, input().split())
papers = list(map(int, input().split()))
sol(N, M, papers)
```
