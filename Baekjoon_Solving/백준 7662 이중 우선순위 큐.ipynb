```python
import heapq
import sys

#input = sys.stdin.readline

T = int(input())

def In(n):
    global total
    heapq.heappush(max_heap , -n)
    heapq.heappush(min_heap , n)
    if n in my_dict:
        my_dict[n] += 1
    else:
        my_dict[n] = 1
    total += 1

def Delete_max():
    global total
    while True:
        num = heapq.heappop(max_heap)
        num = -1 * num
        if my_dict[num] >0:
            my_dict[num] -= 1
            total -= 1
            return

def Delete_min():
    global total
    while True:
        num = heapq.heappop(min_heap)
        num =  num
        if my_dict[num] >0:
            my_dict[num] -= 1
            total -= 1
            return

def last_maxi():
    while True:
        num = heapq.heappop(max_heap)
        num = -1 * num
        if my_dict[num] >0:
            return num

def last_mini():
    while True:
        num = heapq.heappop(min_heap)
        if my_dict[num] >0:
            return num

ans = []
for _ in range(T):
    max_heap = []
    min_heap = []
    my_dict = {}
    total = 0
    N = int(input())

    for _ in range(N):
        comm , num = input().split()
        num = int(num)
        if comm == 'I':
            In(num)
        elif num == 1 and total > 0:
            Delete_max()
        elif num == -1 and total > 0 :
            Delete_min()
    
    if total == 0:
        ans.append(['EMPTY'])
    else:
        ans.append([str(last_maxi()) , str(last_mini()) ] )

for answer in ans:
    print(' '.join(answer))
```
