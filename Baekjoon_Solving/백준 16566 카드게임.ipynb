```python
from bisect import bisect_left
from collections import deque
import sys

def get_data():
    input = sys.stdin.readline
    N, M, K = map(int, input().split())
    min_card_list = list(map(int, input().split()))
    chul_card_list = list(map(int, input().split()))
    return N, M, K, min_card_list, chul_card_list

def find(graph, idx):
    while graph[idx] != idx:
        idx = graph[idx]
    return idx

if __name__ == "__main__":
    N, M, K, min_card_list, chul_card_list= get_data()

    min_card_list.sort()
    graph = {ii : ii for ii in range(M) }
    for ii , chul_card in enumerate(chul_card_list):
        idx = bisect_left(min_card_list, chul_card)
        if min_card_list[idx] == chul_card:
            idx += 1
        idx = find(graph, idx)
        ans = min_card_list[idx]
        print(ans)

        if ii < K-1:
            ## graph 업데이트
            if idx == M -1:
                continue
            else:
                graph[idx] = find(graph, idx+1)
        #print(graph)
```
