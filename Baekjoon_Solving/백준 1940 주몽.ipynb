```python
N = int(input())
M = int(input())
ingredients = list(map( int , input().split() ))
ans = 0

ingredients.sort()

a = 0 ; b = N-1

while True:
    front = ingredients[a]
    back = ingredients[b]

    now = front + back

    if now < M:
        a+=1
    elif now > M:
        b-=1
    else:
        a+=1
        ans+=1
    if a==b:
        break


print(ans)
```
