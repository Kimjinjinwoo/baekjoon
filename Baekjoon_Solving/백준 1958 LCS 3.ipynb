```python
A = input().strip()
B = input().strip()
C = input().strip()

def sol(A, B, C):
    length_A = len(A)
    length_B = len(B)
    length_C = len(C)

    dp = [[0] * length_C for _ in range(length_B)]
    maxi = 0
    for idxA in range(length_A):
        dp_2 = [[0] * length_C for _ in range(length_B)]
        for idxB in range(length_B):
            x = A[idxA]; y = B[idxB]
            if x==y:          
                for idxC in range(length_C):
                    z = C[idxC]
                    if y==z:
                        if idxC == 0 or idxB==0:
                            cnt = 1
                        else:
                            cnt = dp[idxB-1][idxC-1] + 1
                        dp_2[idxB][idxC] = cnt          
        maxi = max(maxi, max([max(row) for row in dp_2]))
        dp = dp_2
        
    print(maxi)

sol(A,B,C)
```
