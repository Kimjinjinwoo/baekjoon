```python
from collections import deque
import sys

def get_input():
    #input = sys.stdin.readline
    raw = list(map(int, input().split()))
    N = raw[0]
    histogram = raw[1:]

    return N, histogram

def update():
    global ans
    global area
    ans = max(ans, area)


def one_step(left_hist, right_hist):
    global ans
    global area
    ## len(left_hist) <= len(right_hist) !!
    area = 0
    left_hist = deque(left_hist)
    right_hist = deque(right_hist)

    ## initial
    min_height = right_hist.popleft()
    area = min_height
    cnt = 1

    ## gogogo
    while(left_hist or right_hist):
        candi = []

        ## 왼쪽거 뽑기
        while(left_hist):
            height = left_hist.pop()
            if height >= min_height:
                cnt += 1
            else:
                left_hist.append(height)
                candi.append(height)
                break
        area = cnt * min_height
        update()

        ## 오른쪽거 뽑기
        while(right_hist):
            height = right_hist.popleft()
            if height >= min_height:
                cnt += 1
            else:
                right_hist.appendleft(height)
                candi.append(height)
                break
        area = cnt * min_height
        update()

        ## min_height 어떻게 update 하지,,,?
        if (candi):
            min_height = max(candi)


## divide_and_conquer 생각해 보자!
def divide_and_conquer(N, hist):
    global ans
    global area
    if N == 1:
        area = hist[0]
        update()
        return hist
    else:
        left_hist = divide_and_conquer(N//2, hist[:N//2]) # deep copy 되는거 생각
        right_hist = divide_and_conquer(N - N//2, hist[N//2:]) # deep copy 되는거 생각

        ## 중간에서 합치는 거
        one_step(left_hist, right_hist)
        return hist

if __name__ == "__main__":
    while(True):
        N, histogram = get_input()
        ## 정지
        if not N:
            break

        ans = 0 # 여기선 전역으로 쓰임.
        area = 0 # 여기선 전역으로 쓰임.

        divide_and_conquer(N, histogram)
        print(ans)
```
