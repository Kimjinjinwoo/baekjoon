```python
A, P = map(int, input().split())

def make_next(num):
    str_num = str(num)
    sum = 0
    for n in str_num:
        sum += int(n) ** P
    return sum

isin = {}
i=1
isin[A] =  i

while True:
    k = make_next(A)
    #이제 반복 됨
    if k in isin:
        ans = isin[k]
        break
    #안 반복
    else:
        i+=1
        isin[k] = i
        A = k

print(ans-1)
```
