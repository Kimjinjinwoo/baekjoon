```python
## 목적지까지 우회하지 않고 바로 간다!
## => 최단 경로를 찾고 g-h 길이 포함되어 있는지 확인하자!
## => 근데 시간이 될랑가??

def get_input():
    N, M, T = map(int, input().split()) ## N = 노드 개수, M = 간선 개수, T = 후보 개
    S, G, H = map(int, input().split()) ## S = start
                                        ## 듀오가 g와 h 사이를 지났음!

    graph = [tuple(map(int, input().split())) for _ in range(M)] ## (A, B, cost)
    des_candi = [int(input()) for _ in range(T)]

    return (N, M, T), (S, G, H), graph, des_candi

if __name__ == "__main__":
    ## Test Case T 번
    for _ in range(int(input())):
        (N, M, T), (S, G, H), graph, des_candi = get_input()
```
