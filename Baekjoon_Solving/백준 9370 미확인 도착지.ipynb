```python
## 목적지까지 우회하지 않고 바로 간다!
## => 최단 경로를 찾고 g-h 길이 포함되어 있는지 확인하자!
## => 근데 시간이 될랑가??

## 그럼 다익스트라를 3번 쓰자!!
## start에서 한 번 g에서 한 번 h에서 한 번해서
## (start - g) + (g - h) + (h - des) == (start - des) 인지 보자!
## 그럼 previous를 저장할 필요도 없겠네

import heapq
from sys import maxsize as inf
import sys

def get_input():
    input = sys.stdin.readline
    N, M, T = map(int, input().split()) ## N = 노드 개수, M = 간선 개수, T = 후보 개
    S, G, H = map(int, input().split()) ## S = start
                                        ## 듀오가 g와 h 사이를 지났음!

    graph = [tuple(map(int, input().split())) for _ in range(M)] ## (A, B, cost)
    des_candi = [int(input()) for _ in range(T)]

    return (N, M, T), (S, G, H), graph, des_candi

def graph2dict():
    dict_graph = { ii : {} for ii in range(1,N+1)}
    for A, B, cost in graph:
        dict_graph[A][B] = cost
        dict_graph[B][A] = cost
    return dict_graph

def dikjestra(start):
    shortest = [inf] * (N+1)
    previous = [-1] * (N+1)

    ## initial
    heap = [ (0, start, -1) ]

    while(heap):
        cost, destination, source = heapq.heappop(heap)
        if shortest[destination] < cost:
            #print(destination, cost)
            #print(shortest)
            continue
        else:
            shortest[destination] = cost
            previous[destination] = source
            ## heap에 넣기
            for next_destination in graph[destination]:
                next_cost = cost + graph[destination][next_destination]
                heapq.heappush(heap, (next_cost, next_destination, destination))
        #print()
    return shortest, previous

def check(now, next):
    if min(now, next) == min(G, H):
        if max(now, next) == max(G, H):
            return True
    return False


def solve():
    ans = []
    des_candi.sort()
    for des in des_candi:
        now = des
        while (previous[now] != -1):
            next = previous[now]
            ## check
            if check(now, next):
                ans.append(str(des))
                break
            now = next
    print(' '.join(ans))
    pass

if __name__ == "__main__":
    ## Test Case T 번
    for _ in range(int(input())):
        (N, M, T), (S, G, H), graph, des_candi = get_input()
        ## 그래프를 딕셔너리 형태로 바꿔서 사용
        graph = graph2dict()
        ## 다익스트라
        shortest, previous = dikjestra(S)
        ## 정답 제출
        solve()
```
