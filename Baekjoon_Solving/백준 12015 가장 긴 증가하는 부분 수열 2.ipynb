```python
from collections import deque
import bisect
N = int(input())
nums = list(map(int, input().split()))
LIS = []

def bis(value):
    start = 0; end = len(LIS) - 1
    while (start <= end):
        #print(start, end)
        mid_idx = (start + end) // 2
        mid_value = LIS[mid_idx]

        if mid_value == value:
            return mid_idx
        elif mid_value < value:
            start = mid_idx + 1
        elif mid_value > value:
            end = mid_idx - 1

    return max(start, end)

def lis():
    LIS.append(nums[0])
    for num in nums:
        if LIS[-1] < num:
            LIS.append(num)
        else:
            LIS[bis(num)] = num
        #print(LIS)
    return len(LIS)

print(lis())
```
