```python
N = int(input())
A,B = map(int, input().split() )
M = int(input())

family = [ [] for _ in range(N+1) ]
#그냥 연결 상태만 보기
for _ in range(M):
    F , S = map(int,input().split())
    family[F].append(S)
    family[S].append(F)

def bfs(family , A,B):
    global N
    visited = [False] * (N+1)
    ans = -1
    isit = False

    queue = [A]
    while queue:
        n = len(queue)
        for i in range(n):
            next = queue.pop(0)
            if next == B:
                queue = []
                isit = True
                break
            if not visited[next]:
                    queue.extend(family[next])
        ans+=1

    if isit:
        return ans
    # 불가능한 경우
    else:
        return -1
print(bfs(family, A , B))
```

```python
N = int(input())
A,B = map(int, input().split() )
M = int(input())


family = [ 0 for _ in range(N+1) ]
# 자식 인덱스에 부모 정보 넣기
for _ in range(M):
    F , S = map(int,input().split())
    family[S] = F



def dfs(W):
    n = 1
    a = family[W]
    parents = [ 0 for _ in range(N+1) ]
    while a:
       parents[a] = n
       a = family[a]
       n+=1
    return parents

A_parents=dfs(A)
B_parents=dfs(B)

ansList = []


for i in range(1,N+1):
    if A_parents[i] and B_parents[i]:
        ansList.append(A_parents[i] + B_parents[i])
     
if A_parents[B]:
    ansList.append( A_parents[B] )
if B_parents[A]:
    ansList.append( B_parents[A])

if ansList:
    print(min(ansList))
else:
    print(-1)
```

```python
#bfs 풀이

from collections import deque
import sys
sys.setrecursionlimit(10 ** 9)

n = int(input())
a,b = map(int , input().split())
m = int(input())
graph = [[] for _ in range(n+1)]
visited = [False] * (n+1)
for i in range(m):
    x,y = map(int , input().split())
    graph[x].append(y)
    graph[y].append(x)

q = deque()
q.append((0,a))

def bfs():
    while q:
        d,x = q.popleft()
        visited[x] = True
        for i in graph[x]:
            if not visited[i]:
                q.append((d+1 , i))
            if i == b:
                return d+1
    return -1

print(bfs())
```
