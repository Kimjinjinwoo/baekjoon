```python
def sol(keywords , bridge):
    keywords = list(keywords)
    N = len(bridge[0])
    dp = dp_init(bridge, keywords[0], N)
    for character in keywords[1:]: 
        dp = next_step(dp, bridge, character, N)
    ans = sum([sum(row) for row in dp])
    print(ans)
    return

def next_step(dp, bridge, character, N):
    next_dp = [[0]*N for _ in range(2)]
    for i in range(2):
        for j in range(N):
            n = dp[i][j]
            if n == 0:
                continue
            for k in range(j+1, N):
                if bridge[1 - i%2][k] == character:
                    next_dp[1-i%2][k] += n
    return next_dp

def dp_init( bridge, character, N ):
    dp = [[0]*N for _ in range(2)]
    for i in range(2):
        for j in range(N):
            if bridge[i][j] == character:
                dp[i][j] = 1      
    return dp

keywords = input()
bridge = [input() for _ in range(2)]
sol(keywords, bridge)
```
