```python
T = int(input())

nums = [ list(map(int, input().split())) for _ in range(T) ]

def gcd(a,b):   
    while True:
        r = a %b
        if r == 0:
            return b
        a,b = b,r

def lcm(a,b):
    a,b = max(a,b) , min(a,b)
    g = gcd(a,b)
    return (a*b // g)

for a,b in nums:
    print(lcm(a,b))
```
