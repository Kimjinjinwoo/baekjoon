```python
anses = []
while True:
    N,M = map(int , input().split())

    if N==0 and M == 0:
        break
    M = M // N
    moneys = list(map(int , input().split()))  
    ans = 0
    for money in moneys:
        if money > M:
            ans += M
        else:
            ans += money

    anses.append(ans)

for ans in anses:
    print(ans)
```
