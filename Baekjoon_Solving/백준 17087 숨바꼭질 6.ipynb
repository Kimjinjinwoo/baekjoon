```python
N , S = map(int , input().split())
siblings = list(map(int , input().split()))

dis = [ abs(loc - S) for loc in siblings ]

def gcd(A,B): # A>B
    while True:
        r = A % B
        if r == 0:
            return B
        A , B = B ,r
A = dis[0]

for i in range(1,N):
    B = dis[i]
    A,B = max(A,B) , min(A,B)
    A = gcd(A,B)

print(A)   
```
