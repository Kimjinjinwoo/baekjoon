```python
import sys
input = sys.stdin.readline

def sol():
    N, Q = map(int,input().split())
    popul = list( map( int, input().split() ))
    cumsum, cumfun = make_cum(N, popul)
    for _ in range(Q):
        l, r = map(int, input().split())
        fun = calc_fun(N, cumsum, cumfun, l, r)
        print(fun)

def make_cum(N, popul):
    cumsum = [0, popul[0]] + [0] * (N-1)
    cumfun = [0]* (N+1)
    for i in range(1, N+1):
        cumsum[i] = cumsum[i-1] + popul[i-1]
        cumfun[i] = cumfun[i-1] + cumsum[i-1] * popul[i-1]
    return cumsum, cumfun

def calc_fun(N, cumsum, cumfun, l, r):
    fun = cumfun[r] - cumfun[l-1]
    fun -= cumsum[l-1] * (cumsum[r] - cumsum[l-1])
    return fun

sol()
```

```python
import sys
#input = sys.stdin.readline

N , Q = map(int , input().split())
teams = list(map(int, input().split()))
divisions = [list(map(int, input().split())) for _ in range(Q)]

def make_cum_sum(N , teams):
    cum_sum = [0]
    for team in teams:
        cum_sum.append(cum_sum[-1] + team)
    return cum_sum

def cal_partial_sum(cum_sum , a , b):
    return cum_sum[b] - cum_sum[a-1]

def cal_total_fun(division , teams , cum_sum):
    l ,r = division
    total_fun = 0
    for a in range(l , r):
        fun = teams[a-1] * cal_partial_sum(cum_sum , a+1 , r)
        #print(a,r,cal_partial_sum(cum_sum , a+1 , r))
        total_fun += fun
    return total_fun

cum_sum = make_cum_sum(N , teams)
for division in divisions:
    print(cal_total_fun(division , teams , cum_sum))
```
