```python
import sys

L1 = list(map(int , input().split()))
L2 = list(map(int , input().split()))

def L1_function(x):
    x1 , y1 , x2 , y2 = L1
    # y = (y1-y2)/(x1-x2) * (x-x1) + y1
    m = (y1-y2)/(x1-x2)
    y = m * (x - x1) + y1
    return y

def L2_function(x):
    x1 , y1 , x2 , y2 = L2
    # y = (y1-y2)/(x1-x2) * (x-x1) + y1
    m = (y1-y2)/(x1-x2)
    y = m * (x - x1) + y1
    return y

def vali(left_ , right_):
    v1 = L1_function(left_) - L2_function(left_) 
    v2 = L1_function(right_) - L2_function(right_) 
    if left_ > right_:
        print(0)
    elif v1 * v2 <= 0:
        print(1)
    else:
        print(0)

def modify():
    global L1 ; global L2
    if L1[0] == L1[2]:
        L1[0] -= 0.25 ; L1[2] += 0.25
    if L2[0] == L2[2]:
        L2[0] -= 0.25 ; L2[2] += 0.25

modify()
left_   =   max(min(L1[0] , L1[2]) , min(L2[0] , L2[2]))
right_  =   min(max(L1[0] , L1[2]) , max(L2[0] , L2[2]))

vali(left_ , right_)
```
