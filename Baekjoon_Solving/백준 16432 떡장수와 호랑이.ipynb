```python
N = int(input())
tteokList = [list(map(int, input().split()))[1:] for _ in range(N)]
```

```python
## 그래프 느낌으로 풀어볼까
def sol(N, tteokList):
    tteokList = [[-1]] + tteokList
    ansList = []
    dp = [0] * (N+1)
    dfs(N, 1, dp, tteokList)
    print(-1)
    return

def dfs(N, n, dp, tteokList):
    before = dp[n-1]
    for tteok in tteokList[n]:
        if tteok != before:
            next_dp = dp[:]
            next_dp[n] = tteok
            if n == N:
                print_dp(dp)
                
            if n < N:
                dfs(N, n+1, next_dp, tteokList)

def print_dp(dp):
    for tteok in dp[1:]:
        print(tteok)


sol(N, tteokList)
```
