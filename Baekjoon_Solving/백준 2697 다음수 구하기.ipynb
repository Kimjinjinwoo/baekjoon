```python
import sys
input = sys.stdin.readline

def sol(str_number):
    num_list = to_numlist(str_number)
    #print(num_list)
    last_sorted_idx = HowLongSort(num_list)
    if last_sorted_idx == 0:
        print("BIGGEST")
        return
    else:
        last_list = findNextNum(num_list, last_sorted_idx)
        for num in last_list:
            print(num,end = '')
        print()
        return

def to_numlist(str_number):
    num_list = []
    for num in str_number:
        num_list.append(int(num))
    return num_list

def HowLongSort(num_list):
    length = len(num_list)
    cnt = 1
    for i in range(length-2, -1, -1):
        now = num_list[i]
        last = num_list[i + 1]
        if last > now:
            break
        else:
            cnt += 1
    return length - cnt

def findNextNum(num_list, last_sorted_idx):
    standard = num_list[last_sorted_idx - 1]
    new_list = num_list[last_sorted_idx:][:]
    for num in new_list[::-1]:
        if num > standard:
            new_list.remove(num)
            new_list.append(standard)
            break
    new_list.sort()
    last_list = num_list[:last_sorted_idx-1] + [num] + new_list
    return last_list

T = int(input())
for _ in range(T):
    number = input().strip()
    sol(number)
```
