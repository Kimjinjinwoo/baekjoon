```python
board = input()

X_s = board.split('.')
dot_s = board.split('X')

ans = ''

def cover_board(X):
    n = len(X)
    if n== 0:
        return ''
    if n % 2 == 1:
        return -1
    else:
        A_n = n // 4
        B_n = (n - A_n * 4) // 2 
        coverd = 'AAAA' * A_n + 'BB' * B_n 
        return coverd

ans = ''
XXX = ''
for x in board:
    if x == '.':
        now = cover_board(XXX)
        if now == -1 :
            break
        ans += now
        ans += '.'
        XXX = ''
    if x == 'X':
        XXX += 'X'
now = cover_board(XXX)
if now == -1:
    print(-1)
else:
    ans += now
    print(ans)
```
