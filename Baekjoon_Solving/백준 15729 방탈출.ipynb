```python
def sol(N, Note):
    bottons = [0] * N
    cnt = 0
    for n in range(N):
        if bottons[n] != Note[n]:
            push(n, N, bottons)
            cnt+=1
    print(cnt)

def push(n, N, bottons):
    bottons[n] = 1 - bottons[n]
    if n < N-1:
        bottons[n+1] = 1 - bottons[n+1]
    if n < N-2:
        bottons[n+2] = 1 - bottons[n+2]

N = int(input())
note = list(map(int, input().split()))
sol(N, note)
```
