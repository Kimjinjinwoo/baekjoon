```python
import heapq
from sys import maxsize
import sys

input = sys.stdin.readline

def my_dijkstra(cave , N):
    visited[0][0] = cave[0][0]
    candi = [[cave[1][0] + visited[0][0] , 1 , 0] , [cave[0][1] + visited[0][0] , 0 , 1]]
    heapq.heapify(candi)

    while candi:
        w , x , y = heapq.heappop(candi)
        if visited[x][y] < w:
            continue
        for dx,dy in zip([0,0,1,-1] , [1,-1,0,0]):
            x_ , y_ = x+dx , y+dy
            if 0<= x_ < N and 0<= y_ < N:
                w_ = w + cave[x_][y_]
                if visited[x_][y_] > w_:
                    visited[x_][y_] = w_
                    heapq.heappush(candi , [w_ , x_ , y_]) 
    return visited[N-1][N-1]

ans = []

while True:
    N = int(input())
    if not N:
        break
    
    cave = [list(map(int,  input().split())) for _ in range(N)]
    visited = [[maxsize] * N for _ in range(N)]

    ans.append(my_dijkstra(cave , N))

for i , num in enumerate(ans):
    print(f'Problem {i+1}: {num}')
```
