```python
def get_charToId():
    C2T = {'(':0, ')':1, '+':2, '-':3, '*':4, '/':5}
    for X in 'QWERTYUIOPASDFGHJKLZXCVBNM':
        C2T[X] = -1
    return C2T

def get_inorder():
    inorder = {0:20, 1:0, 2:1, 3:1, 4:2, 5:2}
    return inorder

def get_outorder():
    outorder = {0:0, 1:0, 2:1, 3:1, 4:2, 5:2}
    return outorder


def main(infix):
    C2T = get_charToId()
    inorder = get_inorder()
    outorder = get_outorder()

    output = []
    stack = []

    for x in infix:
        #print(x, stack)
        id = C2T[x]
        if id == -1:
            output.append(x)
        elif x == ')': ## ')'
            top = stack.pop()
            while(top != '('):
                output.append(top)
                top = stack.pop()
        else:
            if not stack:
                stack.append(x)
                continue
            else:
                top = stack.pop()
                top_order =outorder[C2T[top]]
                x_order = inorder[id]
                if top_order < x_order:
                    stack.append(top)
                    stack.append(x)
                else:
                    while top_order >= x_order:
                        output.append(top)
                        if not stack:
                            break
                        top = stack.pop()
                        top_order =outorder[C2T[top]]
                    if top_order < x_order:
                        stack.append(top)
                    stack.append(x)
    while stack:
        output.append(stack.pop())
    print(''.join(output))

infix = input().strip()
main(infix)
```
