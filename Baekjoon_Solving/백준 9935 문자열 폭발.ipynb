```python
def sol():
    string = input().strip()
    bomb = input().strip()
    
    string = ' ' + string

    len_string = len(string)
    len_bomb = len(bomb)

    dp = [0] * (len_string + 1)

    # 0 번째 칸 비워 둘거임.
    n = 1
    while n < len_string:
        before_ = dp[n-1]
        s = string[n]
        dp[n] = 0

        if s == bomb[0]:
            dp[n] = 1
        
        if s == bomb[before_]:
            dp[n] = before_ + 1

        if dp[n] == len_bomb:
            string = string[:n-len_bomb+1] + string[n+1:]

            n = n - len_bomb + 1
            len_string = len(string)
            continue
        n += 1
    string = string[1:]

    if string:
        print(string)
    else:
        print('FRULA')

sol()  
```
