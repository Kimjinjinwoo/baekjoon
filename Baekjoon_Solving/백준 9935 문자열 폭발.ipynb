```python
import sys
from collections import deque

def sol():
    string = input().strip()
    bomb = input().strip()
    ans = []

    len_bomb = len(bomb)
    last_char = bomb[-1]

    for s in string:
        ans.append(s)
        if s == last_char:
            if ''.join(ans[-len_bomb:]) == bomb:
                for _ in range(len_bomb):
                    ans.pop()

    if ans:
        print(''.join(ans))
    else:
        print('FRULA')

sol()  
```

```python
## 다시 도전~
import sys
from collections import deque

def sol():
    string = input().strip()
    bomb = input().strip()
    
    string = ' ' + string
    ans = deque()

    len_string = len(string)
    len_bomb = len(bomb)

    dp = [0] * (len_string + 1)
    count = 0
    last = 0
    # 0 번째 칸 비워 둘거임.
    n = 1
    for n, s in enumerate(string):
        if not n:
            continue

        #print(n,last)
        #print(ans)
        #print(dp)        
        ans.append(s)
        before_ = dp[last]

        if s == bomb[0]:
            dp[n] = 1 

        if s == bomb[before_]:
            dp[n] = before_ + 1
        
        # 펑 
        if dp[n] == len_bomb:
            for i in range(len_bomb):
                ans.pop()
            count += 1
            last = last - len_bomb + 1
            while dp[last] == len_bomb:
                last = last - len_bomb + 1
        else:
            last = n

    if ans:
        while ans:
            print(ans.popleft(), end = '')
    else:
        print('FRULA')

sol()  
```
