```python
from collections import deque
## 처음에 하나 한쪽에 넣고, 인접한 거 반대로 넘기고 
## 이 과정 반복하는 그런식으로 하는 것이 괜춘 할 듯.

def sol(V, E, edges):
    verteces = [-1] + [0] * (V)
    graph = [ [] for _ in range(V+1)]
    visited = [False] * (len(verteces))
    for edge in edges:
        x, y = edge
        graph[x].append(y)
        graph[y].append(x)
    verteces, visited = bfs(graph, verteces, visited, 1)

    for idx, vertex in enumerate(verteces):
        if vertex == 0:
            verteces, visited = bfs(graph, verteces, visited, idx)
            break
    else:
        print('NO')
        return
    if 0 in verteces:
        print('NO')
        return
    print('YES')

def bfs(graph, verteces, visited, start):
    queue = deque()
    visited[start] = True
    verteces[start] = 1
    queue.append(start)
    while queue:
        x = queue.popleft()
        for next in graph[x]:
            if not visited[next]:
                visited[next] = True
                queue.append(next)
    return verteces, visited

K = int(input())
for _ in range(K):
    V, E = map(int, input().split())
    edges = [tuple(map(int, input().split())) for _ in range(E)]
    sol(V, E, edges)
```
