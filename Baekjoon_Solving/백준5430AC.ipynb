```python
T = int(input())

ansList = []

for _ in range(T):
    cmm = input()
    n = int(input())
    cnt = 0

    try:
        inp = input()
        length = len(inp)
        nums = list(map(int , inp[1:length -1].split(',')))
        #print(nums)
    except:
        nums = []
        

    t = 0
    a=0 ; b = n

    for cm in cmm:
        if cm == 'R':
            t = 1 - t # 0->1 , 1->0
        else:   # cm =='D'
            cnt += 1
            if t: # t==1
                b-=1
            else: # t==0
                a+=1
    
    else:
        if t:# t==1
            an = nums[a:b][::-1]
        else:# t==0
            an = nums[a:b]
    #print(t, a,b)

    if cnt>n:
        ans = "error"
    elif an:
        ans = "["
        ans += str(an[0])
        for a in an[1:]:
            ans += ","
            ans += str(a)
        ans+=']'
    else:
        ans = "[]"
    
    ansList.append(ans)


for ans in ansList:
    print(ans)
```
