```python
from collections import deque

N, M = map(int, input().split())
campus = [input() for _ in range(N)]

visited = [[False] * M for _ in range(N)]

def find_doyeon():
    for ii in range(N):
        for jj in range(M):
            if campus[ii][jj] == 'I':
                return (ii,jj)



doyeon = find_doyeon()
queue = deque([doyeon])
visited[doyeon[0]][doyeon[1]] = True
ans = 0

while(queue):
    x, y = queue.popleft()

    for dx, dy in zip([0,0,1,-1], [1,-1,0,0]):
            nx = x + dx; ny = y + dy
            if (0 <=nx<N and 0<=ny<M):
                if not visited[nx][ny]:
                    visited[nx][ny] = True
                    if campus[nx][ny] == 'X':
                        continue
                    elif campus[nx][ny] == 'P':
                        ans += 1
                    queue.append( (nx, ny) )
if ans:
    print(ans)
else:
    print("TT")
```
