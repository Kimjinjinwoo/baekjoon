```python
test = [1,5]
print(0)

for i in range(10):
    new_test = []
    for n in test:
        new_test.append( int( str(n) + str(1) ) )
        new_test.append( int( str(n) + str(5) ) )

    test = new_test
    cnt = 0
    for n in test:
        if n % 15 == 0:
            #print(n , end=' ')
            cnt+=1
    
    print('\n',cnt)    
```
