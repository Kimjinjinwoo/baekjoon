```python
from collections import deque
import sys
input = sys.stdin.readline

def sol(N, classes):
    classes.sort()
    classes = deque(classes)
    number_of_lecture_room = 0
    while classes:
        number_of_lecture_room += 1
        classes = class_align(classes)
    print(number_of_lecture_room)
    return

def class_align(classes):
    left_classes = deque()
    _, last_end = classes.popleft()
    while classes:
        start, end = classes.popleft()
        if start >= last_end:
            last_end = end
        else:
            left_classes.append( (start, end) )
    return left_classes

N = int(input())
classes = [tuple(map(int, input().split())) for _ in range(N)]
sol(N, classes)
```
