```python
import sys
input = sys.stdin.readline

N = int(input())
power = list(map(int, input().split()))

R = 0 ; B = 0;

def make_RB_power_list():
    R_power = [0] * (N)
    B_power = [0] * (N)

    R_maxi = power[0]; R_power[0] = R_maxi
    B_maxi = power[N-1]; B_power[N-1] = B_maxi

    for i in range(1,N):
        if power[i] > R_maxi:
            R_maxi = power[i]
        if power[N-1 - i] > B_maxi:
            B_maxi = power[N-1-i]

        R_power[i] = R_maxi
        B_power[N-1-i] = B_maxi
    return R_power , B_power

def sol():
    R = 0 ; B = 0;

    for i in range(N-1):
        R_maxi = R_power[i]
        B_maxi = B_power[i+1]
        if R_maxi > B_maxi:
            R += 1
        elif B_maxi > R_maxi:
            B += 1
    
    if R > B:
        print('R')
    elif B > R:
        print('B')
    else:
        print('X')
    #print(R,B)

R_power , B_power = make_RB_power_list()
sol()
```
