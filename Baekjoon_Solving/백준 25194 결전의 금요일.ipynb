```python
N = int(input())
works = list(map(int, input().split()))

works_2 = [0] * 7

#나머지 같은 것들만 모으기
for work in works:
    works_2[ work % 7 ] += 1

works = [0] * 7

for i,work in enumerate(works_2):
    works[i] = min(7 , work)

possible = works

impossible = True
#0인 경우 허용
possible[0] = 10000

def can_do(i,j,k):
    ret = True
    if i==j and j==k:
        if possible[i] <3:
            ret = False
    elif i==j or i==k:
        if possible[i]<2:
            ret = False
    elif j==k:
        if possible[j]<2:
            ret = False
    return ret

#if possible[4]:
#    print("YES")
#    impossible = False

for i in range(7):
    if not impossible:
        break
    for j in range(7):
        if not impossible:
            break
        for k in range(7):
            if not impossible or ( not can_do(i,j,k)):
                continue
            if (i+j+k) % 7 == 4 and possible[i] and possible[j] and possible[k]:    
                print(i,j,k)    
                print("YES")
                impossible = False

if impossible:
    print("NO") 
```

```python
N = int(input())
works = list(map(int, input().split()))

# 목표: 합해서 7로 나눈 나머지 4인 것 찾기

works_2 = [0] * 7

# 나머지 같은것들 모으기
# Ex) 1,8,15,22 -> 1
for work in works:
    works_2[ work % 7 ] += 1

works = [0] * 7

# 많아 봤자 7개 (8개 이상은 의미 없음 (8번 더하면 1번이랑 나머지가 같아서))
for i,work in enumerate(works_2):
    works[i] = min(7 , work)


possible = [[0] for _ in range(7)]

# 같은거 더했을때도 나머지 만 보기
# Ex) 5가 5개 있는 경우는 -> possible[5] = [0,5,3,1,6,4]
for num , work in enumerate(works):
    k = 0
    for i in range(work):
        k += num
        possible[num].append(k % 7)

gogogo = False
#백트래킹으로 다 하기
#이러면 연산 최대 823543(7**7)번 밖에 없음
def rec_ft(n , ans):
    global gogogo
    if n == 7 or gogogo:
        if ans % 7 == 4:
            gogogo = True
            return
        else:
            return
    for num in possible[n]:
        rec_ft(n+1 , ans+num)

#실행
rec_ft(0,0)

#출력
if gogogo:
    print("YES")
else:
    print("NO")
```
