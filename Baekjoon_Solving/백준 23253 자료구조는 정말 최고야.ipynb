```python
import sys
from collections import deque
input = sys.stdin.readline

N, M = map(int , input().split())
dummies = []
for _ in range(M):
    k = int(input())
    dummy = deque(map(int,input().split()))
    dummies.append(dummy)

for dummy in dummies:
    be_sorted = True 
    a = dummy.pop()
    while dummy:
        b = dummy.pop()
        if a > b:
            be_sorted = False
            break
        else:
            a = b
    if not be_sorted:
        break
    
if be_sorted:
    print('Yes')
else:
    print('No')
```
