```python
# 가장 긴 증가하는 수열이네..
# 어케하드랑..

n = int(input())
box_ = list( map( int , input().split() ) )

dp = [1 for _ in range(n)]


for i in range(n):
    for j in range(i+1,n):
        if box_[i] < box_[j]:
            dp[j] = max(dp[j] , dp[i] + 1)
    #print(dp)

print(max(dp)) 
```
