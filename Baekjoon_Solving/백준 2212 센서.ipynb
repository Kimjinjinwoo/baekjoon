```python
def sol(N, K, sencers):
    ## 겹치는거는 의미 없을 듯?
    ## 거리 긴 거를 기준으로 파티션 만들자
    sencers = list(set(sencers))
    sencers.sort()
    N = len(sencers)
    distList = make_distList(N, sencers)
    ans = min_able_range(N, K, sencers, distList)
    print(ans)           
    return

def make_distList(N, sencers):
    distList = [] #(dist, A, B, i)
    for i in range(1, N):
        A, B = sencers[i-1], sencers[i]
        dist = B - A
        distList.append(dist)
    return distList

def min_able_range(N, K, sencers, distList):
    K = min(N,K)
    min_able_range = sencers[-1] - sencers[0]
    distList.sort(key = lambda x:-x)

    for i in range(K-1):
        min_able_range -= distList[i]
    return max(0, min_able_range)

N = int(input())
K = int(input())
sencers = list(map(int, input().split()))

sol(N, K, sencers)    
```
