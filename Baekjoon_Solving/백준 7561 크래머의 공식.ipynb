```python
import sys
input = sys.stdin.readline

T = int(input())

matrices = []

for _ in range(T):
    matrices.append([ list(map(int , input().split()))  for _ in range(3)])

def det(A):
    det_ =  A[0][0] * ( A[1][1] * A[2][2] - A[1][2] * A[2][1]) - A[0][1] * ( A[1][0] * A[2][2] - A[1][2] * A[2][0]) +A[0][2] * ( A[1][0] * A[2][1] - A[1][1] * A[2][0])
    return det_

def change_matrix(A , b , n):
    n -= 1
    A_ = [A[i][:] for i in range(3)]
    A_[0][n] = b[0];
    A_[1][n] = b[1];
    A_[2][n] = b[2];
    return A_

def det_i(A):
    matrix = [A[i][:3] for i in range(3)]
    b = [A[0][3] , A[1][3] , A[2][3]]

    A1 = change_matrix(matrix , b , 1)
    A2 = change_matrix(matrix , b , 2)
    A3 = change_matrix(matrix , b , 3)

    det_A = det(matrix)
    det_A1 = det(A1)
    det_A2 = det(A2)
    det_A3 = det(A3)

    return det_A1 , det_A2 , det_A3 , det_A

def cramer(det_A1 , det_A2 , det_A3 , det_A):
    x1 = det_A1 / det_A
    x2 = det_A2 / det_A
    x3 = det_A3 / det_A

    return modify_value(x1) , modify_value(x2) , modify_value(x3)

def modify_value(x):
    if -0.0005 < x < 0.0005:
        x = 0
    return x

for i,matrix_b in enumerate(matrices):
    det_A1 , det_A2 , det_A3 , det_A = det_i(matrix_b)
    print(det_A1 , det_A2 , det_A3 , det_A)
    if det_A == 0:
        print('No unique solution')
    else:
        x1 , x2 , x3 = cramer(det_A1 , det_A2 , det_A3 , det_A)
        print(f'Unique solution: {x1:.3f} {x2:.3f} {x3:.3f}')
    
    if i != T - 1:
        print()
```
