```python
N = int(input())
food_infos = [ input().split()  for _ in range(N)]
```

```python
def sol(N, food_infos):
    AntNest = {}
    for food_info in food_infos:
        AntNest = update_nest( AntNest, food_info)
    print_nest(AntNest, 0)

def update_nest(AntNest, food_info):
    now_nest = AntNest
    for food in food_info[1:]:
        if food in now_nest.keys():
            now_nest = now_nest[food]
        else:
            now_nest[food] = {}
            now_nest = now_nest[food]
    return AntNest

def print_nest(now_nest , now_depth):
    foods = sorted(now_nest.keys())
    for food in foods:
        print('--' * now_depth + food)
        print_nest(now_nest[food] , now_depth + 1)

N = int(input())
food_infos = [ input().split()  for _ in range(N)]

sol(N ,food_infos)
```
