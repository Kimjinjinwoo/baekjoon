```python
## prim 쓰면 될 듯..?
## prim하고 제일 cost큰 길 없애면 되나..?
import sys
import heapq

def get_data():
    input = sys.stdin.readline
    N, M = map(int, input().split())
    graph_info = [tuple(map(int, input().split())) for _ in range(M)]
    return N, M, graph_info

def make_graph(N, M, graph_info):
    graph = [[]for _ in range(N+1)]
    for A, B, cost in graph_info:
        graph[A].append((cost, B))
        graph[B].append((cost, A))
    return graph

def prim(N, M, graph):
    visited = [False] * (N+1)
    heap = [(0, 1)]
    costs = []

    ### loop
    while heap:
        cost, des = heapq.heappop(heap)
        if not visited[des]:
            visited[des] = True
            costs.append(cost)
            ### heap update
            for des_cost, des_des in graph[des]:
                if not visited[des_des]:
                    heapq.heappush(heap, (des_cost, des_des))

    ### print ans
    print(sum(costs) - max(costs))

if __name__ == "__main__":
    ## 입력받기
    N, M, graph_info = get_data()

    ## graph 만들기
    graph = make_graph(N, M, graph_info) ## [ [], [(cost, des),( cost, des), ...], ... ]

    ## prim?
    prim(N, M, graph)
```
