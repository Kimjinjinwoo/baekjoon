```python
from collections import deque

N ,M = map(int , input().split())
board = [list(map(int, list(input()))) for _ in range(N)]

visited = [[[False]*M for _ in range(N) ]for _ in range(2)]

def can_go(i,j,k,queue):

    go_i = [i,i,i+1,i-1]
    go_j = [j+1,j-1,j,j]

    for ii,jj in zip(go_i,go_j):
        if 0<= ii < N and 0<=jj<M and not visited[k][ii][jj]:
            visited[k][ii][jj] = True
            if board[ii][jj] == 0:
                queue.append([ii,jj,k])
            elif k==0:
                queue.append([ii,jj,1])
    return queue

def bfs():
    global N,M
    ans = 0
    a = [0,0,0]
    queue = deque()
    queue.append(a)
    finish = False
    while queue:
        n = len(queue)
        for _ in range(n):
            i,j,k = queue.popleft()
            if i==N-1 and j==M-1:
                finish = True
                queue = []
                break
            visited[k][i][j] = True
            queue = can_go(i,j,k,queue)
        ans +=1
    if finish:
        return ans
    else:
        return -1

ans = bfs()

print(ans)
```
