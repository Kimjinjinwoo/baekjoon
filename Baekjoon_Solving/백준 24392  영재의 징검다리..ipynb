```python
import sys
inpu = sys.stdin.readline
N, M = map(int , inpu().split())

Bridge = []

for _ in range(N):
    Bridge.append(list(map(int , inpu().split())))

Bridge.reverse()

dp = Bridge[0]
last_row = Bridge[0]

for row in Bridge[1:]:
    dp_2 = [0] * M
    for i in range(M):
        if last_row[i]:
            nums = [i-1 , i , i+1]
            for idx in nums:
                if 0 <= idx < M and row[idx]:
                    dp_2[idx] += dp[i]
    last_row = row[:]
    #print(dp_2)
    dp = dp_2[:]

print(sum(dp) % 1000000007)
```
