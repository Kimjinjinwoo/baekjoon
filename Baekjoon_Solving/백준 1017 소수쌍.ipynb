```python
def Eratosthenes(M = 2000):
    a = [False, False] + [True] * (M-1)

    for ii in range(2, M+1):
        if (a[ii]):

            for jj in range(2*ii, M+1, ii):
                a[jj] = False
    return a

def get_OKprime_map(N, nums, prime_test):
    OKprime_map = [[] for _ in range(N)]

    for ii in range(N):
        for jj in range(ii+1, N):
            num_ii = nums[ii]; num_jj = nums[jj]
            num = num_ii + num_jj
            if prime_test[num]:
                OKprime_map[ii].append(jj)
                OKprime_map[jj].append(ii)

    return OKprime_map

def bimatch(n):
    if visited[n]:
        return False
    visited[n] = True

    for num in OKprime_map[n]:
        if selected[num] == -1 or bimatch(selected[num]):
            selected[num] = n
            return True
    return False


if __name__ == "__main__":
    N = int(input())
    nums = list(map(int, input().split()))

    ans = []
    prime_test = Eratosthenes()
    OKprime_map = get_OKprime_map(N, nums, prime_test)

    for idx in OKprime_map[0]:
        selected = [-1] * N
        selected[0] = idx
        selected[idx] = 0

        for ii in range(N):
            visited = [True] + [False] * (N-1)
            visited[idx] = True
            bimatch(ii)

        if not -1 in selected:
            ans.append(nums[idx])


    if ans:
        #ans = list(set(ans))
        ans.sort()
        for num in ans:
            print(num, end = ' ')
    else:
        print(-1)
```
