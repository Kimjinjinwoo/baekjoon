```python
N = int(input())
nums = list(map(int, input().split()))
```

```python
def Eratosthenes(M = 2000):
    a = [False, False] + [True] * (M-1)

    for ii in range(2, M+1):
        if (a[ii]):

            for jj in range(2*ii, M+1, ii):
                a[jj] = False
    return a

def get_OKprime_map(N, nums, prime_test):
    OKprime_map = [[] for _ in range(N)]

    for ii in range(N):
        for jj in range(ii+1, N):
            num_ii = nums[ii]; num_jj = nums[jj]
            num = num_ii + num_jj
            if prime_test[num]:
                OKprime_map[ii].append(jj)
                OKprime_map[jj].append(ii)

    return OKprime_map

def check(visited, meat_with_first):
    good = False
    for next_idx, next_check in enumerate(visited):
        if not next_check:
            next = next_idx
            break
    else:
        ans.append(nums[meat_with_first])
        return True ## 모두 짝 지었음

    visited[next_idx] = True

    for idx in OKprime_map[next_idx]:
        if not visited[idx]:
            visited[idx] = True
            good = check(visited, meat_with_first)
            visited[idx] = False
            if good:
                break

    visited[next_idx] = False
    if good:
        return True

if __name__ == "__main__":
    N = int(input())
    nums = list(map(int, input().split()))

    ans = []
    prime_test = Eratosthenes()
    OKprime_map = get_OKprime_map(N, nums, prime_test)
    visited = [True] + [False] * (N-1)


    for idx in OKprime_map[0]:
        visited[idx] = True
        check(visited, idx)
        visited[idx] = False

    if ans:
        #ans = list(set(ans))
        ans.sort()
        for num in ans:
            print(num, end = ' ')
    else:
        print(-1)
```
