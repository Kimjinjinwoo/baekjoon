```python
def sol():
    N = int(input())
    M = int(input())

    S = input().strip()

    L = 2*N + 1
    dp = [0] * M
    ans = 0

    if S[0] == 'I':
        dp[0] = 1

    for i in range(M-1):
        if S[i] == 'O':
            if S[i+1] == 'I':
                dp[i+1] = dp[i] + 1
                if dp[i+1] >= L:
                    ans += 1
        if S[i] == 'I':
            if S[i+1] =='O':
                dp[i+1] = dp[i] + 1
            if S[i+1] == 'I':
                dp[i+1] = 1
    print(ans)

sol()
```
