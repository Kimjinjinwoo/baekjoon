```python
from collections import deque
import sys
input = sys.stdin.readline

N = int(input())
bridges = [list(map(int , input().split())) for _ in range(N - 2)]

graph = [ [] for _ in range( N+1 ) ]
label = [0] * (N+1)
label[0] = -1

def make_graph():
    for a , b in bridges:
        graph[a].append(b)
        graph[b].append(a)

def sol():
    queue = deque()
    label[1] = 1
    while graph[1]:
        island = graph[1].pop()
        queue.append(island)

    while queue:
        n = len(queue)
        for i in range(n):
            island = queue.pop()
            if label[island] == 0:
                label[island] = 1
                for new_ in graph[island]:
                    if label[new_] == 0:
                        queue.append(new_)

make_graph()
sol()
print(label.index(1) , label.index(0))
```
