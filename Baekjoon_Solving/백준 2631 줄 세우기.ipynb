```python
def sol(N, nums):
    len_LIS = LIS(N, nums)
    print(N - len_LIS)
    return

def LIS(N, nums):
    dp = [0] * N
    for i in range(N):
        n = nums[i]
        dp[i] = 1
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j]+1)
    return max(dp)

N = int(input())
nums = [int(input()) for _ in range(N)]
sol(N, nums)
```
