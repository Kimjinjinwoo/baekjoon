```python
N = input()
M = int(input())
if M:
    brokens = list(map( int , input().split()))
else:
    brokens = []
N_num = int(N)

def mini_maxi(botton):
    not_brok = []
    for i in range(10):
        if botton[i]:
            not_brok.append(i)
    not_brok.sort()
    if min(not_brok) != 0:
        return min(not_brok) , max(not_brok) , min(not_brok)
    elif len(not_brok) > 1:
        return min(not_brok) , max(not_brok) , not_brok[1]
    else:
        return min(not_brok) , max(not_brok) , -1

def up_number(n):
    up_num = -1
    for i in range(int(n)+1 , 10):
        if botton[i]:
            up_num = i
            break
    return up_num

def down_number(n):
    down_num = -1
    for i in range(int(n)-1 , -1 , -1):
        if botton[i]:
            down_num = i
            break
    return down_num

def check(i , up_num , down_num):
    global upup; global down
    upup = upup[:i] + str(up_num)    + upup[i+1:]
    down = down[:i] + str(down_num)  + down[i+1:]
    up_offset   = int(upup) - int(N)
    down_offset = int(N)    - int(down)

    if up_offset > down_offset:
        return down
    else:
        return upup

def compare(A ,B):
    offset_A = abs(int(A) - int(N))
    offset_B = abs(int(B) - int(N))
    if offset_A < offset_B:
        #print(1)
        return A
    elif offset_A == offset_B:
        if len(A) < len(B):
            #print(2)
            return A
        else:
            #print(3)
            return B
    else:
        #print(4)
        return B
    

def count():
    offset = abs(int(N) - int(ans))
    cnt = len(str(int(ans))) + offset
    return min(cnt , abs(int(N) - 100))


if M<10:
    botton = {}

    for i in range(10):
        botton[i] = 1
    for brok in brokens:
        botton[brok] = 0

    mini , maxi , next_mini = mini_maxi(botton)

    upup = str(mini) * len(N)
    down = str(maxi) * len(N)

    for i,n in enumerate(N):
        n_num = int(n)
        if botton[n_num]:
            upup = upup[:i] + n + upup[i+1:]
            down = down[:i] + n + down[i+1:]
            ans = upup
            continue
        up_num = up_number(n) ; down_num = down_number(n)
        if up_num != -1 and down_num != -1:
            ans = check(i , up_num , down_num)
            break
        elif up_num != -1:
            upup = upup[:i] + str(up_num) + upup[i+1:]
            ans = upup
            if i == 0:
                down = down[1:]
                if down =='':
                    break
                ans = compare(upup , down)
            break
        elif down_num != -1:
            
            down = down[:i] + str(down_num) + down[i+1:]
            ans = down
            if i == 0 and next_mini != -1:
                upup = str(next_mini) + upup[:]
                ans = compare(upup , down)
            break
    zzzz =  str(mini) * (len(N) +1 )
    aaaa =  str(maxi) * (len(N) -1 )

    A = compare(zzzz , aaaa)
    B = compare(ans ,A)
    ans = B

    print(count())
else:
    print(abs(int(N) - 100))
```
