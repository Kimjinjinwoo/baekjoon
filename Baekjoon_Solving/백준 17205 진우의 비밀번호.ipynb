```python
N = int(input())
password = input()

alpha = {}

for i,a in enumerate('abcdefghijklmnopqrstuvwxyz'):
    alpha[a] = i

def sum_of_geometric(n , a = 1 , r = 26):
    seriese = a * (r**n - 1) // ( r-1 )
    return seriese

ans = 0

for i , p in enumerate(password):
    n = N - i
    alpha_value = alpha[p] 
    ans += (sum_of_geometric(n)) * alpha_value + 1 

print(ans)
```
