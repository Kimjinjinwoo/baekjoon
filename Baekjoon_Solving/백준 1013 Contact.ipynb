```python
def verify_01(wave: str, idx: int)->int :
    try:
        n = len(wave)
        assert idx <= n
        if wave[idx] != '0':
            return -1 
        idx += 1
        if wave[idx] != '1':
            return -1
        idx += 1
        return idx
    except:
        return -1

def verify_1001(wave: str, idx: int)->int :
    try:
        n = len(wave)
        assert idx <= n
        ## first
        if wave[idx] != '1':
            return - 1; 
        idx += 1
        if wave[idx] != '0':
            return -1 
        idx += 1
        if wave[idx] != '0':
            return -1
        idx += 1
        ## second
        while (wave[idx] == '0'):
            idx+=1
        ### 마지막 1이 너무너무너무 애매하네네
        return idx+1
    except:
        return -1
```

```python
def analyze(wave: str):
    n = len(wave)
    idx = 0
    while(idx < n):
        if wave[idx] == '0':
            idx = verify_01(wave, idx)
        elif wave[idx] == '1':
            idx = verify_1001(wave, idx)
        if idx == -1:
            print("NO")
            return
        
    print("YES")

def verify_01(wave: str, idx: int)->int :
    try:
        n = len(wave)
        assert idx <= n
        if wave[idx] != '0':
            return -1 
        idx += 1
        if wave[idx] != '1':
            return -1
        idx += 1
        return idx
    except:
        return -1

def verify_1001(wave: str, idx: int)->int :
    try:
        n = len(wave)
        assert idx <= n
        ## first
        if wave[idx] != '1':
            return - 1; 
        idx += 1
        if wave[idx] != '0':
            return -1 
        idx += 1
        if wave[idx] != '0':
            return -1
        idx += 1
        ## second
        while (wave[idx] == '0'):
            idx+=1
        while (wave[idx] == '1'):
            if idx == n-1:
                break
            idx+=1    
        return idx+1
    except:
        return -1

T = int(input())
for _ in range(T):
    wave = input()
    analyze(wave)
```

```python
from collections import deque
def analyze_byStack(wave):
    wave = deque(list(wave))
    success = True
    while(wave and success):
        #print(wave)
        x = wave.popleft()
        # start from 0
        if (x == '0'): 
            if not wave:
                success = False; break;
            x = wave.popleft()
            if (x != '1'):
                success = False
        # start from 1
        elif (x == '1'):
            for _ in range(2):
                if not wave:
                    success = False; wave = deque([]); break;
                x = wave.popleft()
                if (x != '0'):
                    success = False; wave = deque([]); break;
            while (x == '0'):
                if not wave:
                    success = False; wave = deque([]); break;
                x = wave.popleft()
            #print("음? ", x, wave)
            if not wave:
                break
            ## 100+1 까지는 함
            ## 이제 1+
            #print("으악 x : ",x,  wave)
            numOf1 = 0
            while (x == '1'):
                numOf1+=1
                if not wave:
                    break;
                x = wave.popleft()
            ## 지금은 x == 0
            if not wave and x == '1':
                continue; 

            if (not wave or numOf1 == 1) and x == '0':
                wave.appendleft('0')
                continue; 

            x = wave.popleft()
            if (x == '0'):              
                wave.appendleft('0')
                wave.appendleft('0')
                wave.appendleft('1') # 100
                ## 다음은 100
                
            else:
                wave.appendleft('1')
                wave.appendleft('0') # 01
                ## 다음은 01
                

    if success:
        print("YES")
    else:
        print("NO")


T = int(input())
for _ in range(T):
    wave = input().strip()
    analyze_byStack(wave)
```
