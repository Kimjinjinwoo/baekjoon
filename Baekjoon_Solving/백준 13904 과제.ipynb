```python
import sys
input = sys.stdin.readline

def sol(N, assignments):
    day = [[] for _ in range(1001)]
    total = 0
    for assignment in assignments:
        d, score = assignment
        day[d].append(score)
    # heap으로 구현해도 좋겠다.
    lasts = []
    for i, scores_of_the_day in enumerate(day[::-1]):
        if i == 1000:
            break
        lasts.extend(scores_of_the_day)
        if lasts:
            score = max(lasts)
            total += score
            lasts.remove(score)
            #print(score,lasts)
    print(total)

N = int(input())
assignments = [tuple(map(int, input().split())) for _ in range(N)]

sol(N, assignments)
```

```python
import sys
import heapq
#input = sys.stdin.readline

def sol(N, assignments):
    day = [[] for _ in range(1001)]
    total = 0
    for assignment in assignments:
        d, score = assignment
        day[d].append(-1 * score)
    # heap으로 구현해도 좋겠다.
    lasts = []
    for i, scores_of_the_day in enumerate(day[::-1]):
        if i == 1000:
            break
        for score in scores_of_the_day:
            heapq.heappush(lasts, score)
        if lasts:
            score = -1 * heapq.heappop(lasts)
            total += score
    print(total)

N = int(input())
assignments = [tuple(map(int, input().split())) for _ in range(N)]

sol(N, assignments)
```
