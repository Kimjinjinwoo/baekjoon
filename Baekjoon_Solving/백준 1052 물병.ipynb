```python
def sol(N, K):
    ## 그냥 하나하나 해야되나
    number_of_bottle = N
    step = -1
    buying = 0
    while number_of_bottle > K:
        step += 1
        number_of_bottle, buying = Combine_Bottle(step, K, number_of_bottle, buying)
        print(step)
        print(number_of_bottle, buying)
    print(buying)

def Combine_Bottle(step, K, number_of_bottle, buying):
    if number_of_bottle % 2 == 0:
        number_of_bottle = number_of_bottle // 2
        return number_of_bottle, buying
    else:
        if (number_of_bottle // 2 + 1) < K:
            number_of_bottle = K
            return number_of_bottle, buying
        number_of_bottle = (number_of_bottle+1) // 2
        buying += 2 ** step
        return number_of_bottle, buying

N, K = map(int, input().split())
sol(N, K)
```
