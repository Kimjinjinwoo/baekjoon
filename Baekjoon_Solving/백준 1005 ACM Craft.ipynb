```python
import sys

def init_graph(N):
    state_graph = [-1] + [0] * (N)
    dp = [0] * (N + 1)
    connection_graph = [[] for _ in range(N+1)]

    return state_graph, connection_graph, dp

def check_able_building(state_graph):
    able_building = []
    for ii, building in enumerate(state_graph[1:], 1):
        if building == 0:
            able_building.append(ii)
    return able_building


def solve(N, K, build_times, state_graph, connection_graph, dp, W):

    ## 원하는 건물을 완성할 때 까지 실행
    while True:
        ### 0. 목표 건물 완성되는지 확인
        if state_graph[W] == 0:
            print(dp[W] + build_times[W])
            return

        ### 1. 지금 지을 수 있는 건물 확인
        able_building = check_able_building(state_graph)

        ### 2. 가능한 건물들 짓기
        for now_building in able_building:
            state_graph[now_building] = -1

            for next_building in connection_graph[now_building]:
                state_graph[next_building] -= 1
                dp[next_building] = max(dp[now_building] + build_times[now_building], dp[next_building])

if __name__ == "__main__":
    input = sys.stdin.readline
    T = int(input())
    for _ in range(T):
        N, K = map(int, input().split())
        build_times = list(map(int, input().split())) ## len == N
        build_times = [-1] + build_times

        state_graph, connection_graph, dp = init_graph(N)

        for _ in range(K):
            X, Y = map(int, input().split())
            state_graph[Y] += 1
            connection_graph[X].append(Y)
        W = int(input())

        solve(N, K, build_times, state_graph, connection_graph, dp, W)

```
