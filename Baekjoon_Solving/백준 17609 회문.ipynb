```python
def test_palindrom(comm):
    length = len(comm)
    for ii in range(length // 2):
        if comm[ii] != comm[length - 1 - ii]:
            return False
    return True

def test_pseud_palindrom(comm):
    length = len(comm)
    for ii in range(length // 2):
        if comm[ii] != comm[length - 1 - ii]:   
            comm1 = comm[:]
            comm1.pop(length - 1 - ii)
            comm2 = comm[:]
            comm2.pop(ii)
            return test_palindrom(comm1) or test_palindrom(comm2)
```
