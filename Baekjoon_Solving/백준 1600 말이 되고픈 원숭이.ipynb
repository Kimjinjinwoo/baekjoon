```python
import sys


K = int(input())
W,H = map(int , input().split())
board = [list(map(int , input().split())) for _ in range(H)]

visited = [[[False] * W for _ in range(H)] for _ in range(K+1)]

def can_go(state):
    goList = []
    x,y,k = state
    #일반적인 움직임
    go_x = [x+1,x-1,x,x]
    go_y = [y,y,y-1,y+1]

    for xx,yy in zip(go_x  ,go_y):
        if 0<=xx<W and 0<=yy<H:
            if visited[k][yy][xx] or board[yy][xx]:
                continue
            visited[k][yy][xx]=True
            goList.append([xx,yy,k])
    #말 같은 움직임
    if k < K:
        k+=1
        go_x = [x+2,x+2,x-2,x-2,x+1,x+1,x-1,x-1]
        go_y = [y+1,y-1,y+1,y-1,y+2,y-2,y+2,y-2]

        for xx,yy in zip(go_x  ,go_y):
            if 0<=xx<W and 0<=yy<H:
                if visited[k][yy][xx] or board[yy][xx]:
                    continue
                visited[k][yy][xx]=True
                goList.append([xx,yy,k])                
    return goList

def bfs():
    global K,W,H
    a = [0,0,0]
    queue = [a]
    ans = -1
    finish = False

    while queue:
        n = len(queue)
        for _ in range(n):
            now = queue.pop(0)
            x,y,k = now
            visited[k][y][x]=True
            if x==W-1 and y == H-1:
                finish = True
                queue = []
                break
            queue.extend(can_go(now))
        ans +=1
    #print(finish)
    if finish:
        return ans
    else:
        return -1

ans = bfs()
print(ans)
```
