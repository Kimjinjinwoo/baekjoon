```python
expression = input()

# 0. 식이 잘못 되었는지 찾는 함수
    # 연속되는 사칙연산 기호
    # 괄호가 잘 열리고 잘 닫혔는지
# 1. 괄호를 찾아서 푸는 함수
# 2. 사칙연산을 수행하는 함수   
    # eval로 하면 개꿀~

# 재귀로 잘 하면 될 듯.

# 잘못된 식 잡기

#연속되는 사칙연산
opers = '+-*/'
brackets = '()'

# ++ 같은거
def double_oper(expression):
    if (expression[0] in opers) or (expression[-1] in opers):
        return False
    back_op = 0
    for n in expression[1:]:
        now_op = 0
        if n in opers:
            now_op = 1
        if now_op * back_op == 1:
            return False
        back_op = now_op
    return True

# 잘 열고 잘 닫았는 지
def wrong_about_bracket(expression):
    bracket_ = 0
    for n in expression:
        if n == '(':
            bracket_ += 1
        elif n == ')':
            bracket_ -= 1
        
        if bracket_ < 0:
            return False
    if bracket_ > 0:
        return False
    else:
        return True 
# (+3+2) 이런거
def wrong_bracket_oper(expression):
    length = len(expression)
    for i , n in enumerate(expression):
        if n == '(':
            if expression[i+1] in opers:
                return False
        elif n == ')':
            if expression[i-1] in opers:
                return False
    return True
            
        
            

def IsIt_Wrong(expression):
    # 잘 되있으면 True
    double_ = double_oper(expression)
    wrong_bracket = wrong_about_bracket(expression)
    if wrong_bracket:
        wrong_bracket_and_oper = wrong_bracket_oper(expression)
    if double_:
        if wrong_bracket:
            if wrong_bracket_and_oper:
                return True
    return False

if IsIt_Wrong(expression):
    try:
        ans_ = eval(expression)
        if ans_ == None:
            print("ROCK")
        else:
            print(int(ans_))
    except:
        print("ROCK")
else:
    print("ROCK")
```
