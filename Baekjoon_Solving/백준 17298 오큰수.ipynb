```python
from collections import deque
def get_input():
    N = int(input())
    nums = list(map(int, input().split()))
    return N, nums

def ans_update(stack, num):
    while (stack):
        ## stack의 탑을 뽑고
        top, top_idx = stack.pop()
        if top < num: ## 정답 찾았으면 안 넣고
            ans[top_idx] = num
        else: ## 아직 못 찾았으면 stack에 다시 넣기
            stack.append((top, top_idx))
            break

def solve(N, nums):
    stack = deque([(-1,-1)])

    for ii, num in enumerate(nums):
        ## stack의 top을 뽑고
        top, top_idx = stack.pop()
        ## 다시 넣어 줌.
        stack.append((top, top_idx))

        if top < num:
            ans_update(stack, num)

        stack.append((num, ii))


## stack의 작은 인덱스에는
## 큰 수가 들어 있다는게 기본 아이디어
## 즉, stack은 내림차순으로 있을 수 밖에 없음!

if __name__ == "__main__":
    N, nums = get_input()
    ans = [-1] * N
    solve(N, nums)
    ## 이거 예외처리 코드 수정하는거 생각하기
    ans[-1] = -1

    for x in ans:
        print(x,end = ' ')
```
