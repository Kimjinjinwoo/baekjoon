```python
def sol():
    x1, y1 = map(int, input().split())
    x2, y2 = map(int, input().split())
    x3, y3 = map(int, input().split())

    direction = find_direction(x1,x2)   # 수직 = 0; 양의 방향 = 1; 음의 방향 = -1;

    #print(direction)

    if direction == 0:
        direction = find_direction(y1,y2)   # 올라가면 1; 내려가면 -1;
        if direction == 1:
            if x3 == x2:
                print(0)
            elif x3 < x2:   
                print(1)   #반시계
            else:           
                print(-1)  #시계

        if direction == -1:
            if x3 == x2:
                print(0)
            elif x3<x2:
                print(-1)    #시계
            else:
                print(1)    #반시계

    else:
        x3_onto_line = find_x3_onto_line(x1,y1,x2,y2,x3, direction)
        #print(x3_onto_line)
        loc_p3 = find_loc_p3(x3_onto_line, y3 * (x1-x2))  # 겹치면 0; 더 크면 1; 더 작으면 -1;
        if x1-x2 < 0:
            loc_p3 = loc_p3 * -1
        #print(loc_p3)
        if loc_p3 == 0:
            print(0)
        elif direction == 1:
            if loc_p3 == 1:
                print(1)    #반시계
            else:
                print(-1)   #시계
        else:
            if loc_p3 == 1:
                print(-1)   #시계
            else:
                print(1)    #반시계

def find_direction(x1,x2):
    if x1 == x2:
        return 0
    elif x1 < x2:
        return 1
    else:
        return -1
def find_x3_onto_line(x1,y1,x2,y2,x3, direction):
    if direction == 0:
        return 
    else:
        slope = (y1-y2) 
        return slope * (x3 - x1) + y1 * (x1-x2)

def find_loc_p3(x3_onto_line, y3):  # 함수가 find_direction이랑 같기는 한데 헷갈리니까 합치지 말자.
    if x3_onto_line == y3:
        return 0
    elif x3_onto_line < y3:
        return 1
    else:
        return -1

sol()
```

```python
### Use CCW algorithm
def sol():
    x1, y1 = map(int, input().split())
    x2, y2 = map(int, input().split())
    x3, y3 = map(int, input().split())

    vector1 = [x2-x1, y2-y1]
    vector2 = [x3-x1, y3-y1]

    print(CCW(vector1, vector2))

def CCW(vector1 , vector2):
    n = vector1[0] * vector2[1] - vector1[1] * vector2[0] ## cross product( z=0 으로 뒀을때)
    ##### 오른손 법칙 ㄷㄷ
    if n == 0:
        return 0
    elif n > 0:
        return 1
    else:
        return -1

sol()
```
