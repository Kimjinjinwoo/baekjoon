```python
N = int( input() )

nums = list(map(int , input().split()))
before_idx = [-1] * (N+1)
dp = [0] * (N+1)

for n in range(N):
    num = nums[n]
    maxi = -1
    maxi_idx=-1
    for i in range(n):
        if dp[i] > maxi and nums[i] < num:
            maxi = dp[i]
            maxi_idx = i
    dp[n] = maxi + 1
    before_idx[n] = maxi_idx

dp_maxi = -1
dp_maxi_idx = -1

for n in range(N):
    dp_num = dp[n]
    if dp_num > dp_maxi:
        dp_maxi = dp_num
        dp_maxi_idx = n

ansList = []
idx = dp_maxi_idx

while(1):
    ansList.append( nums[idx] )
    idx = before_idx[idx]
    if idx == -1:
        break
    
print(len(ansList))

for i in ansList[::-1]:
    print(i , end = ' ')
```
