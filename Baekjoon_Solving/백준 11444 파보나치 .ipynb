```python
def square_divide_and_conquer(base, exp):
    if exp == 1:
        return base
    
    half_square = square_divide_and_conquer(base, exp//2)
    if exp % 2 == 0: # 짝수라면
        X = half_square * half_square
    else: # 홀수라면
        X = half_square * half_square * base 
    if X > 0:
        X = X % 1000000007
    return X

N = int(input())
root5 = 5 ** (1/2)
a = (1 + root5) / 2
b = (1 - root5) / 2

ans =  (1 / root5 ) * ( square_divide_and_conquer(a, N) - square_divide_and_conquer(b, N) )
print(int(ans))
```
