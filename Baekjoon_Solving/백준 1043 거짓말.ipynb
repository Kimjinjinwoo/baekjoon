```python
from collections import deque

def sol():
    N, M = map(int , input().split())
    know = list(map(int, input().split()))
    if know[0]:
        know = know[1:]
    else:
        know = []

    parties = [list(map(int , input().split())) for _ in range(M)]
    
    # 일단 로 연결되어 있는 파티 구분 하기 (파티션으로)
    # 진실을 아는 사람이 포함된 파티션은?

    graph = make_graph(N, parties)
    partitions = make_partition(N, graph)
    cantTalk = make_cantTalk(N, know, partitions)

    ans = 0
    for party in parties:
        for person in party[1:]:
            if person in cantTalk:
                break
        else:
            ans += 1
    
    print(ans)
    

# 연결 확인
def make_graph(N, parties):
    graph = [[] for _ in range(N+1)]
    for i , party in enumerate(parties):
        for n,person in enumerate(party[1:]):
            for other_person in party[2+n:]:
                graph[person].append(other_person)
                graph[other_person].append(person)
    return graph

def make_partition(N, graph):
    partitions = [0] * (N + 1)
    cnt = 1
    for i in range(1,N+1):
        if partitions[i] != 0:
            continue
        else:
            queue = deque([i])
            how_many = 0
            while queue:
                person = queue.popleft()
                partitions[person] = cnt
                for other_person in graph[person]:
                    if partitions[other_person] == 0:
                        queue.append(other_person)
            cnt += 1
    else:
        return partitions
            
def make_cantTalk(N, know, partitions):
    cantTalks = []
    for know_person in know:
            cantTalks.append(partitions[know_person])
    cantTalks = list(set(cantTalks))

    cantTalkperson = []
    for i in range(1, N+1):
        if partitions[i] in cantTalks:
            cantTalkperson.append(i)
    return cantTalkperson

sol()
```
