```python
import sys
inf = 10e9

def get_input():
    #input = sys.stdin.readline
    N, M = map(int, input().split())
    graph_raw_info = [tuple(map(int, input().split())) for _ in range(M)]
    return N, M, graph_raw_info

## 그래프를 사용하기 좋은 형태로 바꾸기
### dictionary로 구현해 보자 ~
def raw_graph_to_graph(N, graph_raw_info):
    graph = { ii : {} for ii in range(1, N+1)}
    for A, B, weight in graph_raw_info: ## 단방향 그래프
        if B in graph[A]:
            graph[A][B] = min(graph[A][B], weight)
        else:
            graph[A][B] = weight
    return graph

def Bellman_Ford_Algorithm(graph, initial):
    for _ in range(N):
        for start_city in range(1, N+1):
            if initial[start_city] == inf:
                continue
            for des_city in graph[start_city]:
                initial[des_city] = min(initial[des_city],
                                        initial[start_city] + graph[start_city][des_city])
        #print(initial)
    return initial

def check_negative_cycle(first_cycle):
    second_cycle = first_cycle.copy()
    second_cycle = Bellman_Ford_Algorithm(graph, second_cycle)

    #print(first_cycle)
    #print(second_cycle)

    for city in first_cycle:
        if first_cycle[city] > second_cycle[city]:
            return False
    return True


if __name__ == "__main__":
    N, M, graph_raw_info = get_input()
    ## 그래프를 더 좋은 구조로 ~
    graph = raw_graph_to_graph(N, graph_raw_info)

    ## 벨만 포드 사용하기
    initial = {ii : inf for ii in range(1, N+1)}
    initial[1] = 0
    first_cycle = Bellman_Ford_Algorithm(graph, initial)

    ## 음의 사이클이 없는 경우 정답 출력
    if check_negative_cycle(first_cycle):
        for city in range(2, N+1):
            dist = first_cycle[city]
            ## 가는 경로 없음
            if dist == inf:
                print(-1)
            ## 가는 경로 있음
            else:
                print(dist)
    else:
        print(-1)

## 고쳤던 점
### 1. 같은 경로를 나타내는 간선의 가중치가 다르게 여러개 있을 수 있음
### 2. 모든 간선에 대해서 N번 반복해야 함!
### 3. start_city의 initial이 inf인 경우에는 update하지 않음
```
