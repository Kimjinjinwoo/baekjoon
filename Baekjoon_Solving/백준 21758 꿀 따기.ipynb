```python
N = int(input())
honey = list(map(int , input().split()))

# 벌 벌 벌집
def case1(honey , partial_sum , partial_sum_reverse):
    global N
    maxi = 0
    bee_1_origin = partial_sum[-1] - honey[0]
    for i in range(1,N-1):
        bee_1 = bee_1_origin - honey[i]
        bee_2 = partial_sum[-1] - partial_sum[i]
        total_honey = bee_1 + bee_2
        if maxi < total_honey:
            maxi = total_honey
    return maxi

# 벌집 벌 벌
def case2(honey , partial_sum , partial_sum_reverse):
    global N
    maxi = 0
    bee_2_origin = partial_sum_reverse[0] - honey[-1]
    for i in range(1,N-1):
        bee_1 = bee_2_origin - honey[i]
        bee_2 = partial_sum_reverse[0] - partial_sum_reverse[i]
        total_honey = bee_1 + bee_2
        if maxi < total_honey:
            maxi = total_honey
    return maxi

# 벌 벌집 벌
def case3(honey , partial_sum , partial_sum_reverse):
    global N
    maxi = 0
    for i in range(1,N-1):
        bee1 = partial_sum[i] - honey[0]
        bee2 = partial_sum_reverse[i] - honey[-1]
        total_honey = bee1 + bee2
        #print(bee1 , bee2)
        if maxi < total_honey:
            maxi = total_honey
    return maxi

def find_maxi(honey , partial_sum , partial_sum_reverse):
    c1 = case1(honey , partial_sum , partial_sum_reverse)
    c2 = case2(honey , partial_sum , partial_sum_reverse)
    c3 = case3(honey , partial_sum , partial_sum_reverse)
    #print(c1,c2,c3)
    return max(c1, c2 ,c3)

partial_sum = []
last = 0
for h in honey:
    last += h
    partial_sum.append(last)

partial_sum_reverse = []
last = 0
for h in honey[::-1]:
    last += h
    partial_sum_reverse.append(last)
partial_sum_reverse.reverse()

maxi = find_maxi(honey , partial_sum , partial_sum_reverse)

print( maxi )
```
