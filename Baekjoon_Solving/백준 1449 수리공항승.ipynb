```python
def sol(N, L, leak):
    leak.sort()
    cover = [0] * 1001
    for location in leak:
        cover[location] = -1
    print(count_tape(N, L, leak, cover))

def count_tape(N, L, leak, cover):
    cnt = 0
    for location in leak:
        if cover[location] == -1:
            cnt += 1
            cover[location] = 1
            for i in range(1, L):
                if location + i > 1000:
                    break
                if cover[location + i] == -1:
                    cover[location + i] = 1
    return cnt

N, L = map(int, input().split())
leak = list(map(int, input().split()))
sol(N, L, leak)
```
