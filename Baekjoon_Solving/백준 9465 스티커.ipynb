```python
T = int(input())
ansList = []

def find(u,i):
    maxi = 0
    for b in range(i+1):
        for a in range(2):
            if b+a != u+i-1 and b+a != u+i+1 and (a,b)!=(u,i) :
                maxi = max(maxi , dp[a][b])
    return maxi

def cal(n , stikers): 
    dp[0][0],dp[1][0] = stikers[0][0],stikers[1][0]
    for i in range(1,n):
        for u in range(2):
            dp[u][i] = find(u,i) + stikers[u][i]
    #print(dp)

def find_max():
    max_1 = max(dp[0])
    max_2 = max(dp[1])
    return max(max_1 , max_2)

for _ in range(T):
    n = int(input())
    stikers = [list(map(int , input().split())) for _ in range(2)]
    dp = [[0] * n for _ in range(2)]
    cal(n,stikers)
    ansList.append(find_max())


for ans in ansList:
    print(ans)
```

```python
T = int(input())

ansList=[]

for _ in range(T):
    n = int(input())
    stikers = [list(map(int , input().split())) for _ in range(2)]

    dp=[stikers[0][0] , stikers[1][0]]

    for i in range(1,n):
        dp_2 = []
        for j in range(2):
            dp_2.append(max( dp[1-j] + stikers[j][i], dp[j] ))
        dp = dp_2
    ansList.append(max(dp))

for ans in ansList:
    print(ans)
```
