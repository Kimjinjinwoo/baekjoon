```python

def find_possibility(N , K):
    D = N
    D -= (K+1) * (K) // 2
    #print(D)
    if D < 0:
        return False
    else:
        return True

def sol(N,K):
    if find_possibility(N, K):
        ans = calc_max_differnce(N, K)
        print(ans)
        return
    else:
        print(-1)

def calc_max_differnce(N, K):
    diff = K-1

    D = N
    D -= (K+1) * (K) // 2
    if D % K == 0:
        return diff
    else:
        return diff + 1

N, K = map(int, input().split())
sol(N, K)
```
