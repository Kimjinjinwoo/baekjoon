```python
N, S = map(int, input().split())
nums = list(map(int,input().split()))

acc_sum = [0]
for num in nums:
    acc_sum.append(acc_sum[-1] + num)
start, end = 0, 1
ans = N + 2

while (end <= N):
    value = acc_sum[end] - acc_sum[start]

    if value >= S:
        length = end - start
        ans = min(ans, length)
        start += 1

    elif value < S:
        end += 1

    if start == end:
        end += 1
if ans == N+2:
    ans = 0
print(ans)
```
