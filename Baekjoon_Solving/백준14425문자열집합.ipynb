```python
N ,M = map(int, input().split())

sets = [input() for _ in range(N)]

strings = [input() for _ in range(M)]

idx_a = ord('a')
dicts = {}

for i in range(ord('a') - idx_a , ord('z') - idx_a + 1):
    dicts[i] = []

for element in sets:
    idx = ord(element[0]) - idx_a
    dicts[idx].append(element)

ans = 0

for string in strings:
    idx = ord(string[0]) - idx_a
    lst = dicts[idx]
    if string in lst:
        ans += 1

print(ans)
```
