```python
def round(x):
    dot_ = x - int(x)
    if dot_ == 0:
        return int(x)
    elif dot_ < 0.5:
        return int(x)
    else:
        return int(x) + 1

n = int(input())

if n:
    scores = [int(input()) for _ in range(n)]

    scores.sort()

    cut = round( n * 0.15 )
    cutted_scores = scores[cut: n - cut]
    ans = sum(cutted_scores)

    print(round(ans / (n - 2 * cut)))
else:
    print(0)
```
