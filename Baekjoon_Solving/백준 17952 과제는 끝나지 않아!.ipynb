```python
from collections import deque
import sys

input = sys.stdin.readline

N = int(input())

assignments = [list(map(int , input().split())) for _ in range(N)]
my_stack = deque()

score = 0

for assignment in assignments:
    if assignment[0]:
        my_stack.append(assignment)

    if my_stack:
        now_ = my_stack.pop()
        now_[2] -= 1
        if now_[2] == 0:
            score += now_[1]
        else:
            my_stack.append(now_)

print(score)
```
