```python
## 위상정렬로 고고
## 위상정렬에서 불가능한 경우 예외처리 어케 하드라
## 구현 하다보면 생각 나겠지 뭐
from collections import deque
import sys

def get_data():
    input = sys.stdin.readline
    N, M = map(int, input().split())
    pds_pick_list = [list(map(int, input().split())) for _ in range(M)]
    return N, M, pds_pick_list

def init_graph(N, M, pds_pick_list):
    graph = [[] for _ in range(N+1)]
    directed_state = [0] * (N+1)
    for pds_pick in pds_pick_list:
        number_of_singer = pds_pick[0]
        for ii in range(number_of_singer - 1):
            a,b = pds_pick[ii+1: ii+3]
            graph[a].append(b)
            directed_state[b] += 1
    return graph, directed_state

def topological_sort(N, M, graph, directed_state):
    ## initial
    queue = deque()
    visited = [False] * (N+1)
    answer = []

    for singer, state in enumerate(directed_state[1:], 1):
        if not state:
            queue.append(singer)
    while queue:
        singer = queue.popleft()
        visited[singer] = True
        answer.append(singer)

        for next_singer in graph[singer]:
            directed_state[next_singer] -= 1
            if directed_state[next_singer] == 0:
                queue.append(next_singer)
    if len(answer) == N:
        for singer in answer:
            print(singer)
    else:
        print(0)
    return


if __name__ == "__main__":
    ## 데이터 받기
    N, M, pds_pick_list = get_data() ## pds_pick[ [], [N, A,B,C,...], ...]
    ## 그래프 만들기
    graph, directed_state = init_graph(N, M, pds_pick_list)
    ## topological_sort
    topological_sort(N, M, graph, directed_state)

```
