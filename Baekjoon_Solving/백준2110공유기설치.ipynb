```python
N,M = map(int , input().split())

#sys하셈!
houseList = [int(input()) for _ in range(N)]

houseList.sort()

#가능한지 함수 만들어야 할듯
def can(length):
    global M
    k = 1
    router = houseList[0]
    for loc in houseList[1:]:
        if loc - router < length:
            continue
        else:
            k+=1
            router = loc
    if k < M:
        return 0
    else:
        return 1

#최대 거리를 변수로 두고 이진 분류로 찾아볼까
#최소나 최대 일수도 있기에
# a = 최소 -1 ,b = 최대 +1 로 둡니다~
a = 0 ; b = houseList[N-1] - houseList[0]+1
while True:
    length = int((b + a) / 2)
    t = can(length)
    if t: # 가능하면 -->>늘려야 겠지??
        a = length 
    else: # 불가능하면 -->>줄여야 겠지??
        b = length
    #print(a,b)
    if abs(a-b) == 1:
        print(a)
        break
```
