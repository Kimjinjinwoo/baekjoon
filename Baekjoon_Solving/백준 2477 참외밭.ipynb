```python
def sol():
    K = int( input() )
    now_point = (0,0)
    points = []
    for _ in range(6):
        direction, length = map(int, input().split())
        next_point = find_next_point(now_point, direction, length)
        points.append(next_point)
        now_point = next_point
    A = find_Area(points)
    print(A * K)

def find_next_point(now_point, direction, length):
    direc = [(0,0), (1,0), (-1,0), (0, -1), (0,1)]
    direc = direc[direction]
    next_point = (now_point[0] + length * direc[0], now_point[1] + length * direc[1])
    return next_point

def find_Area(points):
    points.sort()
    A1 = abs(points[0][1] - points[1][1]) * abs(points[2][0] - points[0][0])
    A2 = abs(points[-1][1] - points[-2][1]) * abs(points[-2][0] - points[-3][0])
    return A1 + A2

sol()
```
