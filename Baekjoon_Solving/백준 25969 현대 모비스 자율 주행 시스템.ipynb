```python
N, M, K = map(int, input().split())
maze = [list(map(int, input().split())) for _ in range(N)]
patterns = [list(map(int, input().split())) for _ in range(5)]
```

```python
from collections import deque

def analyze_pattern(patterns):
    patternList = []
    for ii in range(5):
        for jj in range(5):
            if patterns[ii][jj] == 1:
                patternList.append( (ii-2, jj-2) )
    return patternList

def main(N, M, K, maze, patterns):
    patternList = analyze_pattern(patterns)
    visited = [ [False]*M for _ in range(N) ]

    queue = deque() ## (x, y, k, length, checkPoint)
    queue.append((0,0,0,0,False))
    #visited[0][0] = True
    while queue:
        print(queue)
        now_x, now_y, now_k, now_length, now_checkPoint= queue.popleft()

        for dx, dy in zip([-1,1,0,0], [0,0,-1,1]):
            next_x = now_x + dx; next_y = now_y + dy
            if 0<= next_x <N and 0<= next_y < M:
                if not visited[next_x][next_y]:
                    if maze[next_x][next_y] == 0:
                        #visited[next_x][next_y] = True
                        if (next_x, next_y) == (N-1, M-1) and now_k == True:
                            print(now_length + 1)
                            return
                        queue.append((next_x, next_y, now_k, now_length + 1, now_checkPoint))
                    elif  maze[next_x][next_y] == 2:
                         #visited[next_x][next_y] = True
                         if (next_x, next_y) == (N-1, M-1) and now_k == True:
                            print(now_length + 1)
                            return
                         queue.append((next_x, next_y, now_k, now_length + 1, True))
        if now_k < K:
            for dx, dy in patternList:
                next_x = now_x + dx; next_y = now_y + dy
                if 0<= next_x <N and 0<= next_y < M:
                    if not visited[next_x][next_y]:
                        if maze[next_x][next_y] == 0:
                            #visited[next_x][next_y] = True
                            if (next_x, next_y) == (N-1, M-1):
                                print(now_length + 1)
                                return
                            queue.append((next_x, next_y, now_k+1, now_length + 1, now_checkPoint))
                        elif  maze[next_x][next_y] == 2:
                            #visited[next_x][next_y] = True
                            if (next_x, next_y) == (N-1, M-1):
                                print(now_length + 1)
                                return
                            queue.append((next_x, next_y, now_k+1, now_length + 1, True))
main(N, M, K, maze, patterns)
```
