```python
from collections import deque
import sys
def sol(N, ocean):
    shark = find_shark(N, ocean)
    shark_size = 2
    step = 0
    eated = 0
    while True:
        #print(shark)
        dist_map = make_dist_map(N, ocean, shark, shark_size)
        best, best_dist = one_step(N, ocean, dist_map, shark_size)
        
        if best != None:
            x,y = shark
            ocean[x][y] = 0
            x,y = best
            ocean[x][y] = 9
            step += best_dist
            shark = best
            eated += 1
            shark_size = calc_shark_size(eated)
        else:
            print(step)
            return
    return

def calc_dist(shark, next_shark):
    dist = 0
    dist += abs(shark[0] - next_shark[0])
    dist += abs(shark[1] - next_shark[1])
    return dist

def calc_shark_size(step):
    size = 2
    for i in range(2,100):
        if step >= i:
            step -= i
            size += 1
        else:
            return size

def find_shark(N, ocean):
    for i in range(N):
        for j in range(N):
            if ocean[i][j] == 9:
                return (i,j)

def make_dist_map(N, ocean, shark, shark_size):
    dist_map = [ [-100] * N for _ in range(N) ]
    x, y = shark
    queue = deque()
    queue.append((x,y,0))
    dist_map[x][y] = 0

    visited = [[False]*N for _ in range(N)]
    visited[x][y] = True

    while queue:
        x,y, dist = queue.popleft()
        dist_map[x][y] = dist
        for dx, dy in zip([-1,0,0,1], [0,-1,1,0]):
            X, Y = x + dx, y + dy
            if 0<= X <N and 0<= Y <N:
                if not visited[X][Y]:
                    if ocean[X][Y] <= shark_size:
                        visited[X][Y] = True
                        queue.append((X,Y, dist + 1))
    return dist_map

def one_step(N, ocean, dist_map, shark_size):
    best = None
    best_dist = 10 ** 6
    for i in range(N):
        for j in range(N):
            size = ocean[i][j]
            dist = dist_map[i][j]
            if size < shark_size and dist != -100 and size != 0 and size != 9: 
                if dist < best_dist:
                    best = (i,j)
                    best_dist = dist
    return best, best_dist

N = int(input())
ocean = [list(map(int, input().split())) for _ in range(N)]
sol(N, ocean)
```
