```python
## distance 갱신할 때 어디서 온건 지 저장해 두면 될 듯.
from sys import maxsize
import sys
from heapq import heappop, heappush
input = sys.stdin.readline
pop, push = heappop, heappush

def get_graph(N, bus_info):
    graph = [[] for _ in range(N+1)]
    for a, b, d in bus_info:
        graph[a].append( (d, b) ) #(dist, position)
    return graph

def Dijkstra(N,graph,  bus_info, start, end):
    distance = [maxsize] * (N+1)
    from_where = [-1] * (N+1)

    distance[start] = 0
    from_where[start] = start

    heap = [(0, start)]
    while heap:
        o_dist, o_position = pop(heap)
        for n_dist, n_position in graph[o_position]:
            dist = o_dist + n_dist
            if dist < distance[n_position]:
                from_where[n_position] = o_position
                distance[n_position] = dist
                if n_position == end:
                    return distance, from_where
                push(heap, (dist, n_position))
    return distance, from_where

def get_route(from_where, start, end):
    route = [end]
    before = end
    while(True):
        if before == start:
            break
        else:
            before = from_where[before]
            route.append(before)
    return route

def main(N, M, bus_info, start, end):
    graph = get_graph(N, bus_info)
    distance, from_where = Dijkstra(N,graph,  bus_info, start, end)
    route = get_route(from_where, start, end)
    print(distance[end])
    print(len(route))
    for point in reversed(route):
        print(point, end = ' ')
    return

N = int(input())
M = int(input())
bus_info = [tuple(map(int, input().split())) for _ in range(M)]
start, end = map(int, input().split())

main(N, M, bus_info, start, end)
```
