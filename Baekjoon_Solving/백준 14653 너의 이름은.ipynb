```python
import sys
input = sys.stdin.readline

N, K, Q = map(int, input().split())

talks = []

for _ in range(K):
    talk = input().split()
    talks.append( (int(talk[0]) , ord(talk[1]) - ord('A')) )

def sol(N,K,Q):
    # len(saw) == N
    # len(talks) == K

    saw = {}
    for i in range(N):
        saw[i] = 0 
    saw[0] = 1
    ans = []

    if talks[Q-1][0] == 0:
        return ans
    for i in range(Q-1  , K):
        if i-1 > 0:
            if talks[i-1][0] == talks[i][0]:
                saw[ talks[i-1][1] ] = 1
        saw[ talks[i][1] ] = 1
    for i in range(N):
        if not saw[i]:
            ans.append(i)

    return ans

ans = sol(N,K,Q)
if ans:
    for an in ans:
        print(chr(65 + an), end = ' ')
else:
    print(-1)
```
