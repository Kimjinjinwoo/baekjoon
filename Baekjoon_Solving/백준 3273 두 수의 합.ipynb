```python
def get_input():
    N = int(input())
    nums = list(map(int, input().split()))
    X = int(input())

    return N, nums, X

## 투 포인터로 구현해보자.
def solve(N, nums ,X):
    front = 0; rear = N-1
    ans = 0

    while (front != rear):
        sum_value = nums[front] + nums[rear]

        if sum_value == X:
            ans += 1
            rear -= 1
        ## 합이 작으면 => 더 큰 값 더해야 함
        elif  sum_value < X:
            front += 1
        ## 합이 큰 경우
        else:
            rear -= 1
    print(ans)


if __name__ == "__main__":
    N, nums, X = get_input()
    ## 정렬
    nums.sort()
    ## 풀기
    solve(N, nums, X)
```
