```python
def Eratosthenes(N):
    a = [False, False] + [True] * (N-1)
    primes = []

    for ii in range(2,N+1):
        if a[ii]:
            primes.append(ii)
            for k in range( ii * 2, N+1, ii ):
                a[k] = False
    return primes

N = int(input())
primes = Eratosthenes(N)

L = len(primes)
acc_sum = [0]
for num in primes:
    acc_sum.append(acc_sum[-1] + num)

start, end = 0, 0
ans = 0

while( end <= L ):
    value = acc_sum[end] - acc_sum[start]

    if value == N:
        ans += 1

    if value <= N:
        end += 1
    elif value > N:
        start += 1

    if end == start:
        end += 1
print(ans)
```
