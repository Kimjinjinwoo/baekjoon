```python
import sys 
input = sys.stdin.readline

def sol(K, MarbleList, N, WeightList):
    ## 만들수 있는 무개들을 찾고
    ## 구슬 무게 + 만들 수 있는 무게1 = 만들 수 있는 무게2
    ## 또는 구슬 무게 = 만들 수 있는 무게 찾자.
    WeightList.sort()
    maxsize = sum(WeightList)
    dp = [1] + [0] * (maxsize)

    for Weight in WeightList:
        next_dp = [0] * (maxsize + 1)
        
        for ii in range(maxsize + 1):
            if dp[ii]:
                next_dp[ii] = 1
                if ii + Weight < maxsize + 1:
                    next_dp[ii+Weight] = 1
        dp = next_dp
    
    #print(dp)
    for Marble in MarbleList:
        if maxsize < Marble:
            print('N')
            continue
        for ii in range(maxsize + 1 - Marble):
            if dp[ii] and dp[ii+Marble]:
                print('Y', end = ' ')
                break
        else:
            print('N', end = ' ')
                          
    return 

N = int(input())
WeightList = list(map(int, input().split()))
K = int(input())
MarbleList = list(map(int, input().split()))

sol(K, MarbleList, N, WeightList)
```
