```python
#import sys
#input = sys.stdin.readline

from collections import deque

M,N,H = map(int , input().split())
queue = deque()

box = [[ list(map(int , input().split() ) )  for _ in range(N) ] for _ in range(H)]

def first():
    for z in range(H):
        for y in range(N):
            for x in range(M):
                tomato = box[z][y][x]
                if tomato == 1:
                    queue.append( [ x , y , z ] )

def end():
    for z in range(H):
        for y in range(N):
            for x in range(M):
                tomato = box[z][y][x]
                if tomato == 0:
                    return 0
    return 1

first()

dxL = [1,-1,0,0,0,0]
dyL = [0,0,1,-1,0,0]
dzL = [0,0,0,0,1,-1]
cnt = -1


while (queue):
    cnt+=1
    n = len(queue)
    for _ in range(n):
        x,y,z = queue.popleft()

        for dx,dy,dz in zip(dxL , dyL , dzL):
            nx = x + dx; ny = y +dy ; nz = z + dz
            if 0<=nx<M and 0<=ny<N and 0<=nz<H:
                if box[nz][ny][nx] == 0:
                    queue.append([nx,ny,nz])
                    box[nz][ny][nx] = 1
    #print(queue)

t = end()

if t:
    print(cnt)            
else:
    print(-1)
```
