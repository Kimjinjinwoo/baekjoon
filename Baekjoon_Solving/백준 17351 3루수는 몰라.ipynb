```python
from collections import deque
import sys
input = sys.stdin.readline

def sol():
    global ground , connected , count , visited
    N = int(input())
    ground = [input().strip() for _ in range(N)]
    connected = [ [0]*N for _ in range(N)]
    count = [ [0]*N for _ in range(N)]
    visited = [ [0]*N for _ in range(N)]
    # bfs로 해볼까나~
    
    bfs(N)
    print(count[N-1][N-1])

def bfs(N):
    queue = deque()
    queue.append( (0,0) )
    visited[0][0] = 1
    if ground[0][0] == 'M':
        connected[0][0] = 1
    while queue:
        i,j = queue.popleft( )
        dx = [i, i+1]
        dy = [j+1 , j]
        for x , y in zip(dx , dy):
            if 0<=x<N and 0<=y<N:
                verify(i,j,x,y)
                if not visited[x][y]:
                    queue.append((x,y))
                    visited[x][y] = 1

def verify(i,j , x,y):
    pre = ground[i][j]
    now = ground[x][y]

    mola = 'MOLA'
    ### case 1
    ### case 1-1
    if pre == 'L' and connected[i][j] ==1:
        if now == 'A':
            if count[x][y] < count[i][j] + 1:
                connected[x][y] = 1
                count[x][y] = count[i][j] + 1

    ### case 1-2
    if count[x][y] < count[i][j]:
        count[x][y] = count[i][j]
        #connected[x][y] = connected[i][j]
        if now == 'M':
            connected[x][y] = 1
            return 
        else:
            connected[x][y] = 0
            
        if connected[i][j] == 1:
            for i in range(3):
                if pre == mola[i]:
                    if now == mola[i+1]:
                        connected[x][y] = 1
                        return
            connected[x][y] = 0

    ### case 2
    if count[x][y] == count[i][j]:
        if connected[x][y] == 0:
            if now == 'M':
                connected[x][y] = 1
                return
            if connected[i][j] == 1:
                for i in range(3):
                    if pre == mola[i]:
                        if now == mola[i+1]:
                            connected[x][y] = 1
                            
    # 다른 길에서 max를 고르는데 connect는 다른길에서 된 경우는 우야지..
    #1. count + 1인거
    #1-1. MOLA완성
    #1-2. 그냥 count가 큼
    #2. count는 같지만 connected 인거
    #3. 그냥
    
sol()
```
