```python
### DFS로 그래프 몇개 나오는지 확인
### 그냥 visited를 하나 새로 만들자 ~

import sys

def get_data():
    input = sys.stdin.readline
    N, M = map(int, input().split())
    graph = [input().strip() for _ in range(N)]

    return N, M, graph

def get_char2direction():
    char2direction = {}
    char2direction['U'] = (-1, 0)
    char2direction['D'] = (1, 0)
    char2direction['L'] = (0, -1)
    char2direction['R'] = (0, 1)

    return char2direction

def find_start(N, M, visited):
    for ii in range(N):
        for jj in range(M):
            if not visited[ii][jj]:
                return (ii, jj)
    ## 모든 곳을 방문했으면 None을 반환
    return None

def or_mapping(N, M, map_A, map_B):
    for ii in range(N):
        for jj in range(M):
            if not map_A[ii][jj]:
                map_A[ii][jj] = map_B[ii][jj]
    return map_A


def onestep(N, M, graph, visited, start):
    now_visited =  [[False] * M for _ in range(N)]

    y, x = start
    now_visited[y][x] = True
    char2direction = get_char2direction()
    ### visited를 만날 때 까지
    while True:
        dy, dx = char2direction[graph[y][x]]
        next_y = y + dy; next_x = x + dx
        if 0<= next_y < N and 0<= next_x < M:
            ### 기존 그래프
            if visited[next_y][next_x] == True:
                visited = or_mapping(N, M, visited, now_visited)
                return False

            ### 새로운 그래프
            if now_visited[next_y][next_x] == True:
                visited = or_mapping(N, M, visited, now_visited)
                return True

            now_visited[next_y][next_x] = True
            x, y = next_x, next_y
        else:
            visited = or_mapping(N, M, visited, now_visited)
            return True



if __name__ == "__main__":
    ### 입력 받기
    N, M, graph = get_data()
    visited = [[False] * M for _ in range(N)]
    ans = 0

    for ii in range(N):
        for jj in range(M):
            ### 시작점 찾고
            if visited[ii][jj] == 0:
                start = (ii, jj)
            else:
                continue
            ### dfs 한 바퀴 돌리기
            new = onestep(N, M, graph, visited, start)
            if new:
                ans += 1

    print(ans)
```

```python
### DFS로 그래프 몇개 나오는지 확인
### 고유 index를 사용하자

import sys

def get_data():
    input = sys.stdin.readline
    N, M = map(int, input().split())
    graph = [input().strip() for _ in range(N)]

    return N, M, graph

def get_char2direction():
    char2direction = {}
    char2direction['U'] = (-1, 0)
    char2direction['D'] = (1, 0)
    char2direction['L'] = (0, -1)
    char2direction['R'] = (0, 1)

    return char2direction

def find_start(N, M, visited):
    for ii in range(N):
        for jj in range(M):
            if not visited[ii][jj]:
                return (ii, jj)
    ## 모든 곳을 방문했으면 None을 반환
    return None

def or_mapping(N, M, map_A, map_B):
    for ii in range(N):
        for jj in range(M):
            if not map_A[ii][jj]:
                map_A[ii][jj] = map_B[ii][jj]
    return map_A


def onestep(N, M, graph, visited, start, idx):
    y, x = start
    char2direction = get_char2direction()
    ### visited를 만날 때 까지
    while True:
        dy, dx = char2direction[graph[y][x]]
        next_y = y + dy; next_x = x + dx
        if 0<= next_y < N and 0<= next_x < M:
            ### 새로운 그래프
            if visited[next_y][next_x] == idx:
                return True

            ### 기존 그래프
            if visited[next_y][next_x] > 0:
                return False

            visited[next_y][next_x] = idx
            x, y = next_x, next_y
        else:
            return True



if __name__ == "__main__":
    ### 입력 받기
    N, M, graph = get_data()
    visited = [[False] * M for _ in range(N)]
    ans = 0
    idx = 0

    for ii in range(N):
        for jj in range(M):
            ### 시작점 찾고
            if visited[ii][jj] == 0:
                start = (ii, jj)
            else:
                continue

            idx += 1

            ### dfs 한 바퀴 돌리기
            new = onestep(N, M, graph, visited, start, idx)
            if new:
                ans += 1

    print(ans)
```
