```python
from sys import maxsize

def sol(N, houses):
    ### 집 위치의 누적합 다 구하고
    ### 위치 1부터 거리 구하면 될 듯
    start, end = min(houses), max(houses)
    cumsum, count = make_cumsum_count(N, houses, start, end)
    #print(cumsum)
    #print(count)
    _, best_loc = find_best(N, cumsum, count, start, end)
    print(best_loc)
    return

def make_cumsum_count(N, houses, start, end):
    cumsum = [0] * (end+1)
    count = [0] * (end+1)
    for house in houses:
        cumsum[house] += house
        count[house] += 1
    for i in range(1,end+1):
        cumsum[i] += cumsum[i-1] 
        count[i] += count[i-1] 
    return cumsum, count

def find_best(N, cumsum, count, start, end):
    best = maxsize
    for i in range(start, end+1):
        dist = now_dist(N, cumsum, count, i)
        #print(i, dist)
        if dist < best:
            best_loc = i
            best = dist
    return best, best_loc

def now_dist(N, cumsum, count, i):
    dist = 0
    if i > 0:
        dist += (i*count[i-1] - cumsum[i-1])
    if i < len(cumsum) - 1:
        dist += ( (cumsum[-1] - cumsum[i]) - (N-count[i]) *i )
    return dist

N = int(input())
houses = list(map(int, input().split()))
sol(N, houses)
```
