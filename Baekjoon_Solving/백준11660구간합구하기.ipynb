```python
N , M = map(int , input().split())

locList= []

board = [list(map(int , input().split())) for _ in range(N)]

for _ in range(M):
    loc = list(map(int , input().split()))
    locList.append(loc)

partialSum = [0]*N
partialSum[0] = sum(board[0])

def mkitPartialSum():
    k = partialSum[0]
    for i in range(1 , N):
        k += sum(board[i])
        partialSum[i] = k

mkitPartialSum()

for loc in locList:
    x1,y1,x2,y2 = loc
    x1-=1 ; x2-=1 ;y1-=1 ;y2-=1
    ans = partialSum[x2] - partialSum[x1]

    for i in range(y1, N):
        ans += board[x1][i]
    for i in range(y2+1 , N):
        ans -= board[x2][i]

    print(ans)
```

```python
N , M = map(int , input().split())

locList= []

board = [list(map(int , input().split())) for _ in range(N)]

for _ in range(M):
    loc = list(map(int , input().split()))
    locList.append(loc)

partialSum = [[0]*(N+1) for _ in range(N+1)]
partialSum[1][1] = board[1][1]
for i in range(1 , N+1):
    partialSum[i][1] = partialSum[i-1][1] + board[i-1][0]
    partialSum[1][i] = partialSum[1][i-1] + board[0][i-1]

def mkitPartialSum():
    for i in range(2,N+1):
        for j in range(2,N+1):
            partialSum[i][j] = partialSum[i-1][j] + partialSum[i][j-1] - partialSum[i-1][j-1] + board[i-1][j-1]

mkitPartialSum()

for loc in locList:
    x1,y1,x2,y2 = loc
    ans = partialSum[x2][y2] - partialSum[x2][y1-1] - partialSum[x1-1][y2] + partialSum[x1-1][y1-1]
    print(ans)
```
