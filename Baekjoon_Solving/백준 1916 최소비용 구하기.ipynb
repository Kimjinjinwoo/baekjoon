```python
import sys
import heapq
import collections

inf = 100001

N = int(input())
M = int(input())
distance = [inf] * N
visited = [False] * N
graph = collections.defaultdict(list)

for _ in range(M):
    v1,v2,w = map(int, sys.stdin.readline().split())
    v1 -= 1; v2 -= 1
    graph[v1].append([w,v2])

Start , End = map(int , input().split())

def my_dijkstra(Start , End):
    S = 0
    candi = graph[Start]
    visited[Start] = True
    distance[Start] = 0
    heapq.heapify( candi )
    while candi:
        w , v2 = heapq.heappop( candi )
        if visited[v2]:
            continue
        if distance[v2] < w:
            continue
        for i in graph[v2]:
            cost = w + i[0]
            if cost < distance[i[1]] and not visited[i[1]]:
                distance[i[1]] = cost
                heapq.heappush(candi , [cost , i[1]])

my_dijkstra(Start-1 , End-1)
print(distance[End-1])
```

```python
import heapq
from sys import maxsize
import sys

#input=sys.stdin.readline

n = int(input())
m = int(input())

graph = [[] for _ in range(n + 1)]
visited = [maxsize] * ( n + 1 )
for _ in range(m):
    a,b,c = map(int , input().split())
    graph[a].append((c,b))

start, end = map(int , input().split())

def dijkstra(x):
    pq = []
    heapq.heappush(pq , (0,x))
    visited[x] = 0

    while pq:
        d ,x = heapq.heappop(pq)

        if visited[x] < d:  # 더 긴 거리면은 시도 할 이유가 없음.
            continue

        for nw, nx in graph[x]:
            nd = d + nw

            if visited[nx] > nd:
                heapq.heappush(pq , (nd, nx))
                visited[nx] = nd

dijkstra(start)

print(visited[end])
```
