```python
N, M = map(int, input().split())
graphs = [list(map(int, input().split()))[1:] for _ in range(N)]

selected = [-1] * (M+1)

def bimatch(N):
    if visited[N]:
        return False
    visited[N] = True

    for num in graphs[N]:
        if selected[num] == -1 or bimatch(selected[num]):
            selected[num] = N
            return True
    return False

for ii in range(N):
    visited = [False] * N
    bimatch(ii)

print(sum(list(map(lambda x : 1 if x >= 0 else 0,  selected ) )))
```
