```python
def sol():
    global town_map, roads, partition, graphs, trip_plan
    N = int(input())
    M = int(input())
    roads = [list(map(int, input().split())) for _ in range(N)]
    trip_plan = list(map(int, input().split()))

    graphs = [[] for _ in range(N)]
    link_town(N)

    partition = [0] * N
    make_partition(N)

    verify_plan()

def link_town(N):
    for i, road_from_i in enumerate(roads):
        for j, road in enumerate(road_from_i):
            if road == 1:
                graphs[i].append(j)
                graphs[j].append(i)
    
def make_partition(N):
    label = 1
    for i in range(N):
        if partition[i] == 0:   #아직 미분류
            partition[i] = label
            dfs_for_link(i, N, label)
            label += 1

def dfs_for_link(i, N, label):
    for can_go in graphs[i]:
        if partition[can_go] == 0:
            partition[can_go] = label
            dfs_for_link(can_go, N, label)
    return

def verify_plan():
    ans = 'YES'
    part = partition[ trip_plan[0]-1 ]
    for location in trip_plan:
        if partition[location-1] != part:
            ans = 'NO'
            break
    print(ans)

sol()
```
