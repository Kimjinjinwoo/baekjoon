```python
from sys import maxsize

N = int(input())
bright = list(map(int, input().split()))
turn = list(map(int, input().split()))

original_bright = 0
offset = [0] * (N+1)

for i in range(N):
    if turn[i]:
        original_bright += bright[i]

for i in range(1, N+1):
    offset[i] = offset[i-1]
    if turn[i-1]: #1
        offset[i] += -1 * bright[i-1]
    else:
        offset[i] += bright[i-1]

dp = [offset[1] , offset[1]]
dp_idx = [1 , 1]

max_offset = 0

for i in range(1 , N+1):

    if offset[i]<dp[0]:
        dp[0] = offset[i]
        dp_idx[0] = i

    if offset[i]>dp[1]:
        dp[1] = offset[i]
        dp_idx[1] = i
        now_offset = dp[1] - dp[0]
        if now_offset > max_offset:
            max_offset = now_offset
    #print(dp)

print(max_offset + original_bright)
```
