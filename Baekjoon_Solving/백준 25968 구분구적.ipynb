```python
### polynomial에 exp작은 순으로 저장하기
def sol(n, poly, k):
    if calc_poly(-32 , poly) > 0:
        poly = list(map( lambda x: -x , poly))
    posi_point = find_posi_point(poly)
    left_point = binary_search_left(poly, -32, posi_point)
    right_point = binary_search_right(poly, posi_point, 32)
    ans = integral_by_part(poly, left_point, right_point, k)
    #print(posi_point, left_point, right_point)   
    print(ans)



def calc_poly(x : float, poly : list) -> int:
    ans = 0;
    #print(poly)
    for ii, coef in enumerate(poly):
        ans += coef * (x ** ( (ii) * 2) )
    return ans

def find_posi_point(poly : list) -> int:
    x = -32
    while (x < 32):
        if calc_poly(x, poly) > 0:
            return x
        x += 0.1

def binary_search_left(poly, start, end):
    mid = (start + end) / 2
    for _ in range(10**2):
        mid = (start + end) / 2
        x = calc_poly(mid, poly)
        if (x > 0):
            end = mid
        elif (x < 0):
            start = mid
        else:
            return mid
    return mid

def binary_search_right(poly, start, end):
    mid = (start + end) 
    for _ in range(10**2):
        mid = (start + end) / 2
        x = calc_poly(mid, poly)
        if (x < 0):
            end = mid
        elif (x > 0):
            start = mid
        else:
            return mid
    return mid

def integral_by_part(poly, left_point, right_point, k):
    k = min(k, 10 ** 5)
    d = (right_point - left_point) / k
    S = 0
    for _ in range(k):
        S += d * calc_poly(left_point + d/2, poly)
        left_point += d
    return S

n = int(input())
poly = list(map(float, input().split()))[::-1]
k = int(input())
sol(n, poly, k)
```

```python
def power(a, n):
    if n == 0:
        return 1
    
    x = power(a, n//2)

    if n % 2 == 0:
        return x * x
    
    else:
        return x * x * a
t = time.time()
for i in range(10**5):
    power(i, 100)
print(time.time() - t)
```
