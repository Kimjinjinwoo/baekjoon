```python
import sys

poly = sys.stdin.readline().strip()

def integ(term):
    degree = 1 if term[-1] == 'x' else 0
    if degree == 0:
        a = int(term)
        if a==1:
            return [1,'+x'] if len(terms) == 2 else [1,'x']
        elif a==-1:
            return [1,'-x']
        elif a==0:
            return [0,'']
        else:
            if a > 0:
                return [1, term + 'x']
            else:
                return [1, term + 'x']
    else:
        a = int(term[:-1]) // 2
        if a==1:
            return [2, 'xx']
        elif a==-1:
            return [2, '-xx']
        else:
            if a > 0:
                return [2, str(a) + 'xx']
            else:
                return [2, str(a) + 'xx']

def poly_split(poly):
    if 'x' in poly:
        x_idx = poly.index('x')
        #1. 항 2개
        if x_idx != len(poly) - 1:
            return [poly[:x_idx+1] , poly[x_idx+1:]]
        #2. 1차항만
        else:
            return [poly]  
    #3. 상수항만
    else:
        return [poly]

terms = poly_split(poly)

ans = []

for term in terms:
    ans.append(integ(term))
if poly != '0' and poly != '-0':
    ans.append([0,'+W'])
else:
    ans.append([0,'W'])

for term in ans:
    print(term[1] , end='')
```
