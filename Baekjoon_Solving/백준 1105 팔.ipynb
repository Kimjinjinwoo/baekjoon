```python
def sol(L, R):
    Ln, Rn = len(L) , len(R)
    if Ln != Rn:
        print(0)
        return
    else:
        LList, RList = str2numList(L,R)
        ans = not_eight_possible(LList, RList)
        print(ans)
    return

def str2numList(L, R):
    LList = []
    RList = []
    for n in L:
        LList.append(int(n))
    for n in R:
        RList.append(int(n))
    return LList, RList

def not_eight_possible(LList, RList):
    n = len(LList)
    ans = 0
    for i in range(n):
        if LList[i] == RList[i]:
            if LList[i] == 8:
                ans += 1
        else:
            break
    return ans

L, R = input().split()
sol(L,R)
```
