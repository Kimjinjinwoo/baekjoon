```python
import sys
input = sys.stdin.readline

N,H  = map(int , input().split())
from_bottom = [0 for _ in range(H+1)]
from_top = [0 for _ in range(H+1)]
obstacles = [int(input()) for _ in range(N)]

def when_even(n):
    n = 2 * n
    height = obstacles[n]
    from_bottom[height] += 1

def when_odd(n):
    n = 2 * n + 1
    height = obstacles[n]
    from_top[height] += 1

def plus_correspond(n):
    when_even(n)
    when_odd(n)

for n in range(N//2):
    plus_correspond(n)

to_break = [0 for _ in range(H+1)]
sum_from_bottom = [0 for _ in range(H+1)]
sum_from_top = [0 for _ in range(H+1)]

# 아래서 세기
sum_from_bottom[H] = 0
for h in range(H-1 , 0 , -1):
    sum_from_bottom[h] += from_bottom[h] + sum_from_bottom[h+1]

# 위에서 세기
sum_from_top[H] = 0
for h in range(H-1 , 0 , -1):
    sum_from_top[h] = from_top[h] + sum_from_top[h+1]
 
# 합하기
for h in range(1,H+1):
    to_break[h] = sum_from_bottom[h] + sum_from_top[H+1-h]

M = min(to_break[1:])
N = to_break[1:].count(M)
print(M , N)    
```
