```python
import heapq
import sys
input = sys.stdin.readline

def sol(N, problems):
    dates = [[] for _ in range(N+1)]
    heap = [] 
    total = 0
    for problem in problems:
        deadLine, noodle = problem
        dates[deadLine].append(noodle)
    for date in dates[:0:-1]:
        if not heap and not date:
            continue
        if date:
            for noodle in date:
                heapq.heappush(heap, noodle * -1)
        now = heapq.heappop(heap) * -1
        #print(now)
        total += now
    print(total)

N = int(input())
problems = [tuple(map(int, input().split())) for _ in range(N)]
sol(N, problems)
```
