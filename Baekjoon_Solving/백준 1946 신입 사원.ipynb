```python
import sys
input = sys.stdin.readline

T = int(input())
for _ in range(T):
    N = int(input())
    applicants = [list(map(int, input().split())) for _ in range(N)]
    
    applicants.sort()
    mini = applicants[0][1]
    cnt = N
    for applicant in applicants[1:]:
        rate = applicant[1]
        if rate < mini:
            mini = rate
        else:
            cnt -= 1
    print(cnt)
```
