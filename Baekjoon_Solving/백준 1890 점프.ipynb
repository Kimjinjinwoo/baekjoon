```python
N = int(input())

board = [list(map(int , input().split())) for _ in range(N)]

dp = [[0]*N for _ in range(N)]
dp[0][0] = 1
did_go = [[False]* N for _ in range(N)]

def can_go(i,j):
    length = board[i][j]
    count = dp[i][j]
    a = length
    if a == 0 or (i ==N-1 and j ==N-1):
        return
    if i+a <= N-1:
        dp[i+a][j] += count
        can_go(i+a , j)     
    if j+a <= N-1:
        dp[i][j+a] += count
        can_go(i , j+a)

can_go(0,0)
print(dp[N-1][N-1])
```

```python
N = int(input())

board = [list(map(int , input().split())) for _ in range(N)]

dp = [[0]*N for _ in range(N)]
dp[0][0] = 1
did_go = [[False] *N for _ in range(N)]

queue = [ [0,0] ]

while queue:
    n = len(queue)
    for _ in range(n):
        i,j = queue.pop(0)
        did_go[i][j] = True
        print(queue , dp)
        count = dp[i][j]
        length = board[i][j]
        if length == 0 or (i ==N-1 and j == N-1):
            continue
        if i+length <= N-1:
            dp[i+length][j] += count
            if not did_go[i+length][j]:
                queue.append([i+length,j])   
        if j+length <= N-1:
            dp[i][j+length] += count
            if not did_go[i][j+length]:
                queue.append([i,j+length])

print(dp[N-1][N-1])
```
