```python
import sys
input = sys.stdin.readline

def sol(N, logs):
    logs.sort()
    logs = my_stretch(N,logs)
    ans = calc_difficulty(N,logs)
    print(ans)

def my_stretch(N,logs):
    stretched = [logs[0]] + [0] * (N-1)
    for n in range(1, N):
        if n % 2 == 1:
            stretched[ n//2 + 1 ] = logs[n]
        else:
            stretched[N - n//2] = logs[n]
    return stretched

def calc_difficulty(N, logs):
    difficulty = abs(logs[N-1] - logs[0])
    for n in range(1,N-1):
        diff1 = abs(logs[n] - logs[n-1])
        diff2 = abs(logs[n+1] - logs[n])   
        difficulty = max(difficulty, diff1, diff2 )
    return difficulty

T = int(input())
for _ in range(T):
    N = int(input())
    logs = list(map(int, input().split()))
    sol(N, logs)
```
