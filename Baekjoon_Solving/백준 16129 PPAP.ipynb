```python
#재귀 느낌으로 다가 해볼까
def sol(string):
    if string == 'P' or string == 'PPAP':
        print('PPAP')
        return
    len_string = len(string)
    idx = 0
    try:
        if string[idx] == 'P':
            idx = PorPPAP(string, idx+1 , len_string)
        else:
            print('NP')
            return

        if string[idx] == 'P':
            idx = PorPPAP(string, idx+1 , len_string)
        else:
            print('NP')
            return

        if string[idx] == 'A':
            idx += 1
        else:
            print('NP')
            return

        if string[idx] == 'P':
            idx = PorPPAP(string, idx+1 , len_string)
        else:
            print('NP')
            return
    except:
        print('NP')
        return
    if idx == len_string:
        print('PPAP')
    else:
        print('NP')     

def PorPPAP(string, origin_idx , len_string):
    idx = origin_idx

    for i,s in enumerate('PAP'):
        if idx >= len_string:
            break

        if s == 'P':
            if string[idx] == 'P':
                if idx + 3 + (3-i) < len_string:
                    idx = PorPPAP(string, idx+1, len_string)
                else:
                    idx += 1
            else:
                break
        if s == 'A':
            if string[idx] == 'A':
                idx += 1
            else:
                break
    else:
        return idx
    return origin_idx
    

sol(input().strip())
```

```python
def sol(string):
    stack = ['*','*','*']
    for s in string:
        stack.append(s)
        if s == 'P':
            if ''.join(stack[-4:]) == 'PPAP':
                for _ in range(4):
                    stack.pop()
                stack.append('P')
    if ''.join(stack) == '***P':
        print('PPAP')
    else:
        print('NP')

sol(input().strip())
```
