```python
import sys
input = sys.stdin.readline

N = int(input())
nums = list(map(int, input().split()))

start, end = 0, N-1

min_value = nums[-1] + nums[0]
min_start, min_end = start, end

while start < end:
    value = nums[start] + nums[end]

    if abs(value) < abs(min_value):
        min_value = value
        min_start, min_end = start, end

    if value > 0:
        end -= 1

    elif value < 0:
        start += 1

    elif value == 0:
        break

print(nums[min_start], nums[min_end])
```
