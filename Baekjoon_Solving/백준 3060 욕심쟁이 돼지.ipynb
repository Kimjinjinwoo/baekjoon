```python
T = int(input())
ans = []

for _ in range(T):
    maxi = int(input())
    foods = list(map(int , input().split() ))
    give = 0
    cnt = 1

    while True:
        give = sum(foods)
        if give > maxi:
            break
        cnt += 1
        foods_2 = []
        for i in range(6):
            a = (i - 1) % 6
            b = (i + 1) % 6
            c = (i + 3) % 6

            foods_2.append(foods[i] + foods[a] +foods[c] + foods[b] )
        foods = foods_2[:]
    ans.append(cnt)

for j in ans:
    print(j)
```
