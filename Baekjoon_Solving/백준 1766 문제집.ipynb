```python
from collections import deque
import heapq
import sys
input = sys.stdin.readline

def sol(N, M, CompList):
    ## 위상 정렬 하는데 한 스텝 마다 sort해서 넣어 주면 안될랑가
    ## => heap 구조
    graph = make_graph(N, CompList)
    AdjacencyMatrix = make_AdjacencyMatrix(N, CompList)
    result = []
    queue = []
    queue = AddTo_queue(queue, AdjacencyMatrix)
    while queue:
        tmpList = []
        x = heapq.heappop(queue)
        result.append(x)
        for y in graph[x]:
            AdjacencyMatrix[y] -= 1
            if AdjacencyMatrix[y] == 0:
                heapq.heappush(queue, y)
    print(' '.join(list(map(str, result))))
    return

def make_graph(N, CompList):
    graph = [[] for _ in range(N+1)]
    for Comp in CompList:
        A, B = Comp
        graph[A].append(B)
    return graph

def make_AdjacencyMatrix(N, CompList):
    AdjacencyMatrix = [-1] + [0] * N 
    for Comp in CompList:
        A, B = Comp
        AdjacencyMatrix[B] += 1
    return AdjacencyMatrix

def AddTo_queue(queue, AdjacencyMatrix):
    for idx in range(1, N+1):
        if AdjacencyMatrix[idx] == 0:
            heapq.heappush(queue, idx)
    #tmp_queue.sort()
    return queue

N, M = map(int,input().split())
CompList = [tuple(map(int, input().split())) for _ in range(M)]
sol(N, M, CompList)
```
