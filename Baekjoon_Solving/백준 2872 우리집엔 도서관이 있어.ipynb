```python
import sys
input = sys.stdin.readline

N = int(input())
books = [int(input()) for _ in range(N)]

def book_sort(N, books):
    cnt = 1
    for i,book in enumerate(books):
        if book == N:
            #print(i)
            for next_book in books[:i][::-1]:
                #print(next_book)
                if N-cnt == next_book:
                    cnt += 1
                    
    print(N - cnt)

book_sort(N,books)
```
