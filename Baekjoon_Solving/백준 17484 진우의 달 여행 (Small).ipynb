```python
N,M = map( int , input().split() )

space = [ list(map(int , input().split())) for _ in range(N)]
dp = space[0][:]

def one_go(n):
    global dp
    if n == N:
        return
    go = space[n][:]
    dp_next = []
    for i in range(M):
        money_now = go[i]
        back = [i-1 , i , i+1]
        pocket = []
        for a in back:
            if a==-1 or a==M:
                continue
            else:
                pocket.append(dp[a] + money_now)
        dp_next.append(min(pocket))
    dp = dp_next[:]
    print(dp)
    one_go(n+1)

one_go(1)

print(min(dp))

```
