```python


def sol():
    global last_time, direction
    N, T = map(int, input().split())
    commands = []
    for _ in range(N):
        time, direc = input().split()
        time = int(time)
        direc = 1 if direc == 'right' else 2
        ## 1이 오른쪽 2가 왼쪽
        commands.append( (time, direc) )
    ## 0은 마지막
    commands.append( (T,0) )
    location = [0,0]
    direction = (1,0,0)
    last_time = 0
    for comm in commands:
        direction, location = move(comm, direction, location , last_time)
        last_time = comm[0]
    print(location[0], location[1])

def move(comm, direction, location, last_time):
    #print(comm, direction, location)
    directions = ( (1,0,0),
                   (0,-1,1),
                   (-1,0,2),
                   (0,1,3))
    
    time = comm[0] - last_time
    location[0] += direction[0] * time
    location[1] += direction[1] * time

    if comm[1] == 0: # comm[1] == 1
        pass
    elif comm[1] == 1:
        next_direc = (direction[2] + 1) % 4
        direction = directions[next_direc]
    else:
        next_direc = (direction[2] -1) % 4
        direction = directions[next_direc]
    return  direction, location
sol()
```
