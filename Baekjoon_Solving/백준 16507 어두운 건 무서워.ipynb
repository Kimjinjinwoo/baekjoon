```python
import sys
input = sys.stdin.readline

R , C , Q = map(int , input().split())

photo = [list(map(int, input().split())) for _ in range(R)]
how = [list(map(int ,input().split())) for _ in range(Q)]

cumsum = [[0] * (C+1) for _ in range(R+1)]

def make_cumsum():
    for r in range(1,R+1):
        for c in range(1,C+1):
            cumsum[r][c] = photo[r-1][c-1] + cumsum[r-1][c] + cumsum[r][c-1] - cumsum[r-1][c-1]

def cal_bright(verteces):
    r1,c1,r2,c2 = verteces
    S = (r2-r1+1) * (c2-c1+1)
    bright = 0
    bright += cumsum[r2][c2]
    bright -= cumsum[r1-1][c2]
    bright -= cumsum[r2][c1-1]
    bright += cumsum[r1-1][c1-1]
    print(bright // S)

make_cumsum()
for verteces in how:
    cal_bright(verteces)
```
