```python
import heapq
from sys import maxsize
import sys

input = sys.stdin.readline

V , E = map(int , input().split())
Start = int(input())
visited = [maxsize] * (V+1)
graph = [[] for _ in range(V+1)]

for _ in range(E):
    u , v , w = map(int , input().split())
    graph[u].append([w , v])

def my_dijkstra(start):
    candi = []
    heapq.heappush(candi , [0,start])
    visited[start] = 0

    while candi:
        w , v = heapq.heappop(candi)

        if visited[v] < w:
            continue
        
        for nw , nv in graph[v]:
            nw = nw + w
            if visited[nv] > nw:
                heapq.heappush(candi , [nw, nv])
                visited[nv] = nw

my_dijkstra(Start)

for d in visited[1:]:
    if d == maxsize:
        print('INF')
    else:
        print(d)
    
```
