```python
import sys
input = sys.stdin.readline

def sol(N, K, stuffList):
    dp = [0] * (K+1)
    for weight, value in stuffList:
        for ii in range(K, -1, -1 ):
            if ii - weight< 0:
                break
            dp[ii] = max(dp[ii-weight] + value, dp[ii])
    print(max(dp))

N, K = map(int, input().split())
stuffList = [tuple(map(int, input().split())) for _ in range(N)]

sol(N, K, stuffList)
```
