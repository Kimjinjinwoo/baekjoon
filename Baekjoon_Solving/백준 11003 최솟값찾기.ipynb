```python
import heapq
from collections import deque

N,L = map(int, input().split())
nums = list(map(int, input().split()))
mini = nums[0]

nums = deque(nums)
heap = []
loc = {}

idx = 1
ans = []
while(nums):
    num = nums.popleft()
    loc[num] = idx

    if num < mini:
        mini = num
        loc = {num : idx}
        heap = []
    else:
        if loc[mini] + L <= N:
            heapq.heappush(heap, num)

    if (loc[mini] >= idx - L + 1 ):
        D = mini
    else:
        while(heap):
            D = heapq.heappop(heap)
            if (loc[D] < idx - L + 1 ):
                continue
            else:
                break
    mini = D
    ans.append(D)
    idx += 1

print(' '.join(map(str, ans)))
```
