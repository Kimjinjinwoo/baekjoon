```python
import sys
N = int(input())

base_list = [{} for _ in range(26)]
max = 0

for _ in range(N):
    name = sys.stdin.readline().lower()
    c = name[0]
    a = ord(c) - 97
    dic = base_list[a]


    try:
        dic[name] += 1
        cnt = dic[name]
        if cnt > max:
            max = cnt
            ans = name
        elif cnt == max:
            str_list = [ans,name]
            str_list.sort()
            ans = str_list[0]
    except:
        dic[name] = 1
        if max == 0:
            max = 1
            cnt = 1
            ans = name
        elif max == 1:
            str_list = [ans,name]
            str_list.sort()
            ans = str_list[0]

print(ans)
```
