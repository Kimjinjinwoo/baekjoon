```python
import heapq
import sys
input = sys.stdin.readline

## heapq로 제일 작은거 두개 씩 합치면 되지 않나?
def sol(N, cards):
    number_of_deck = N
    cmp_count = 0
    heapq.heapify(cards)
    while number_of_deck > 1:
        smallest = heapq.heappop(cards)
        next_smallest = heapq.heappop(cards)
        sum_cards = smallest + next_smallest
        cmp_count += sum_cards
        heapq.heappush(cards, sum_cards)
        number_of_deck -= 1
    print(cmp_count)
    return

N = int(input())
cards = [int(input()) for _ in range(N)]
sol(N, cards)
```
