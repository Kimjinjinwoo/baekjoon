```python
def sol(N, nums):
    dp = [[None, None] for _ in range(N)]
    dp[0] = [nums[0], 0]
    last_dp = [nums[0]]
    for n, num in enumerate(nums):
        if n == 0:
            continue
        dp[n][0] = max(dp[n-1][0] + nums[n], nums[n])
        dp[n][1] = max(dp[n-1][0], dp[n-1][1] + nums[n]) 
        last_dp.append(max(dp[n]))
    print(max(last_dp))

N = int(input())
nums = list(map(int, input().split()))

sol(N, nums)
```
