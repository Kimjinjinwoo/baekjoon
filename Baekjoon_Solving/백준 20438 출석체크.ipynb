```python
import sys
input = sys.stdin.readline

N , K , Q , M = map( int , input().split() )
sleep = list(map(int,input().split()))
code_firsts = list(map(int,input().split()))
sections = [list(map(int , input().split())) for _ in range(M)]

students = [0] * (N+3)
cumsum = [0] * (N+3)

def check_attendance(N):
    for code_first in code_firsts:
        if code_first in sleep:
            continue
        for n in range(code_first , N+3 , code_first):
            if n in sleep:
                continue
            else:
                students[n] = 1

def make_cum_sum():
    for i in range(1 , N+3):
        cumsum[i] = cumsum[i-1] + students[i]

check_attendance(N)
make_cum_sum()

for a,b in sections:
    print(b - (a-1) - (cumsum[b] - cumsum[a-1]))
```
