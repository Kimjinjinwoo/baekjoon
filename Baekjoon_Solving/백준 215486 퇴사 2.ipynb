```python
import sys
input = sys.stdin.readline

def sol(N, schedule):
    dp = [0] * N
    for i in range(N):
        t, price = schedule[i]
        if i+t-1 < N:
            dp[i+t-1] = max(dp[i+t-1], dp[i-1]+price)
        if i == 0:
            continue
        dp[i] = max(dp[i], dp[i-1])
    print(dp[i])

N = int(input())
schedule = [tuple(map(int, input().split())) for _ in range(N)]
sol(N, schedule)
```
