```python
N = int(input())
M = int(input())
ans = 1

dp_nc = [0,1,1] + [0] * 40
dp_c =[0,0,1] + [0] *40

for i in range(3,41):
    dp_nc[i] = dp_nc[i-1] + dp_c[i-1]
    dp_c[i] = dp_nc[i-1]

fixed = [int(input()) for _ in range(M)] #sort 되있음.
back = 0


for i in fixed:
    now = i - back - 1
    if not now:
        back = i
        continue
    ans *= (dp_c[now] + dp_nc[now])
    back = i
if N-back:
    ans *= (dp_nc[N - back] + dp_c[N-back])

print(ans)
```
