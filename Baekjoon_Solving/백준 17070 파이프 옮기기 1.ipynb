```python
## (location, state)로 BFS합시다.
## state 0: 가로, 1: 세로, 2: 대각선
from collections import deque

def next_horizon(N, house, location):
    x, y = location
    next_x = x; next_y = y+1
    if 0<= next_x < N and 0<= next_y < N:
        if house[next_x][next_y] != 1:
            return ( (next_x, next_y), 0)
    return None

def next_vertical(N, house, location):
    x, y = location
    next_x = x+1; next_y = y
    if 0<= next_x < N and 0<= next_y < N:
        if house[next_x][next_y] != 1:
            return ( (next_x, next_y), 1)
    return None

def next_diagonal(N, house, location):
    x, y = location
    next_x = x+1; next_y = y+1
    if 0<= next_x < N and 0<= next_y < N:
        if house[next_x][next_y] != 1 and house[next_x][next_y - 1] != 1 and house[next_x - 1][next_y] != 1:
            return ( (next_x, next_y), 2)
    return None



def main(N, house):
    ans = 0
    queue = deque()
    queue.append( ((0, 1),  0) )

    while queue:
        #print(queue)
        location, state = queue.popleft()
        if location == (N-1, N-1):
            ans += 1
        if state == 0:
            next = next_horizon(N, house, location)
            if next:
                if next[0] == (N-1, N-1):
                    ans += 1
                else:
                    queue.append(next)
            next = next_diagonal(N, house, location)
            if next:
                if next[0] == (N-1, N-1):
                    ans += 1
                else:
                    queue.append(next)

        elif state == 1:
            next = next_vertical(N, house, location)
            if next:
                if next[0] == (N-1, N-1):
                    ans += 1
                else:
                    queue.append(next)
            next = next_diagonal(N, house, location)
            if next:
                if next[0] == (N-1, N-1):
                    ans += 1
                else:
                    queue.append(next)

        elif state == 2:
            next = next_horizon(N, house, location)
            if next:
                if next[0] == (N-1, N-1):
                    ans += 1
                else:
                    queue.append(next)
            next = next_vertical(N, house, location)
            if next:
                if next[0] == (N-1, N-1):
                    ans += 1
                else:
                    queue.append(next)
            next = next_diagonal(N, house, location)
            if next:
                if next[0] == (N-1, N-1):
                    ans += 1
                else:
                    queue.append(next)
    print(ans)

N = int(input())
house = [list(map(int, input().split())) for _ in range(N)]

main(N, house)
```

```python
## dp로 합시다....
## state 0: 가로, 1: 세로, 2: 대각선
from collections import deque

def next_vertical(N, house, location):
    x, y = location
    next_x = x+1; next_y = y
    if 0<= next_x < N and 0<= next_y < N:
        if house[next_x][next_y] != 1:
            return next_y, 1
    return None, None

def next_diagonal(N, house, location):
    x, y = location
    next_x = x+1; next_y = y+1
    if 0<= next_x < N and 0<= next_y < N:
        if house[next_x][next_y] != 1 and house[next_x][next_y - 1] != 1 and house[next_x - 1][next_y] != 1:
            return next_y, 2
    return None, None

def get_horizon_cum(N, house, dp, xx):
    ## state 0: 가로, 1: 세로, 2: 대각선
    new_dp = [pixel[:] for pixel in dp]

    for ii in range(1, N):
        if house[xx][ii] != 1:
            new_dp[0][ii] = new_dp[2][ii-1] + new_dp[0][ii-1]
        else:
            new_dp[0][ii] = 0
    return new_dp


def main(N, house):
    dp = [[0]*N for _ in range(3)]
    dp[0][0] = 1
    dp = get_horizon_cum(N, house, dp, 0)
    dp[0][0] = 0
    #print(dp) ##########################
    for xx in range(N-1):
        #print(dp)
        next_dp = [[0]*N for _ in range(3)]
        for yy in range(N):
            for state in range(3):
                location = (xx, yy)
                count = dp[state][yy]

                if state == 0:
                    next_y, next_state = next_diagonal(N, house, location)
                    if next_y:
                        next_dp[next_state][next_y] += count

                elif state == 1:
                    next_y, next_state = next_vertical(N, house, location)
                    if next_y:
                        next_dp[next_state][next_y] += count

                    next_y, next_state = next_diagonal(N, house, location)
                    if next_y:
                        next_dp[next_state][next_y] += count

                elif state == 2:
                    next_y, next_state = next_vertical(N, house, location)
                    if next_y:
                        next_dp[next_state][next_y] += count

                    next_y, next_state = next_diagonal(N, house, location)
                    if next_y:
                        next_dp[next_state][next_y] += count

        next_dp = get_horizon_cum(N, house, next_dp, xx+1)
        dp = [pixel[:] for pixel in next_dp]
        #print(dp)
    print(dp[0][N-1] + dp[1][N-1] + dp[2][N-1])


N = int(input())
house = [list(map(int, input().split())) for _ in range(N)]

main(N, house)           
```
