```python
N = int(input())
nums = list(map( int , input().split() ) )

dp = [0] * N

for i in range(N):
    sum = nums[i]
    mini = 0
    for j in range(i+1 , N):
        num = nums[j]
        sum += num
        dp[j] = max(dp[j] , sum , sum - mini )
        if num < 0:
            mini = min(num , mini )

print( max(dp) )
```
