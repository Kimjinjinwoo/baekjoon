```python
import sys
#input = sys.stdin.readline

N = int(input())
snows = list(map(int,input().split()))
snows.sort()

def clean(N , snows):
    time = 0
    now_big = snows.pop()
    while snows:
        next_big = snows.pop()
        if now_big > next_big:
            now_big -= next_big
            time += next_big
        else:
            time += now_big
            now_big = next_big - now_big

    return time
time = clean(N , snows)
if time >1440:
    print(-1)
else:
    print(time)
```

```python
import sys
import heapq
input = sys.stdin.readline

N = int(input())
snows = list(map(int,input().split()))

def minus(x):
    return -x

def clean(N , snow_heap):
    time = 0
    #now_big = -1 *  heapq.heappop(snow_heap)
    while N > 1:
        #print(N , snow_heap)
        now_big = -1 *  heapq.heappop(snow_heap)
        next_big = -1 *  heapq.heappop(snow_heap)
        
        time += 1

        now_big -= 1
        next_big -= 1

        if now_big:
            heapq.heappush( snow_heap , -1 * now_big)
        else:
            N -= 1
        if next_big:
            heapq.heappush( snow_heap , -1 * next_big)
        else:
            N -= 1
    if N == 1:
        time += -1 * heapq.heappop(snow_heap)

    return time

snow_heap = list(map(minus , snows))
heapq.heapify(snow_heap)
time = clean(N , snow_heap)


if time >1440:
    print(-1)
else:
    print(time)
```
