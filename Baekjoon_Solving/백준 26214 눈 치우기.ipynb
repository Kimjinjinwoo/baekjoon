```python
import sys
#input = sys.stdin.readline

N = int(input())
snows = list(map(int,input().split()))
snows.sort()

def clean(N , snows):
    time = 0
    now_big = snows.pop()
    while snows:
        next_big = snows.pop()
        if now_big > next_big:
            now_big -= next_big
            time += next_big
        else:
            time += now_big
            now_big = next_big - now_big

    return time
time = clean(N , snows)
if time >1440:
    print(-1)
else:
    print(time)
```
