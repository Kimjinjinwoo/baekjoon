```python
from sys import maxsize
import sys
input = sys.stdin.readline

def init_distance(N, bus_info):
    distance = [[maxsize]*(N+1) for _ in range(N+1)]
    for i in range(N+1):
        distance[i][i] = 0
    for a, b, dist in bus_info:
        distance[a][b] = min(distance[a][b], dist)
    return distance

def floyd(N, distance):
    for x in range(1,N+1):
        for a in range(1,N+1):
            if x == a:
                continue
            for b in range(1, N+1):
                if b == x or a == b:
                    continue
                distance[a][b] = min(distance[a][b] , distance[a][x] + distance[x][b])
    return distance

def main(N, M, bus_info):
    distance = init_distance(N, bus_info)
    floyd(N, distance)

    for ii in range(1,N+1):
        for jj in range(1,N+1):
            dist = distance[ii][jj]
            if dist == maxsize:
                dist = 0
            print(dist, end = '')
            if jj < N:
                print(' ', end = '')
        print()
    return

N = int(input())
M = int(input())
bus_info = [tuple(map(int, input().split())) for _ in range(M)]

main(N, M, bus_info)
```
