```python
from sys import maxsize

A,B = input().split()
A_n = len(A); B_n = len(B)

if B_n > A_n:
    A,B = B,A
    A_n , B_n = B_n , A_n

mini = maxsize

for n in range(A_n - B_n + 1):
    offset = 0
    for i in range(B_n):
        if A[i + n] != B[i]:
            offset += 1
    if mini > offset:
        mini = offset

print(mini)
```
