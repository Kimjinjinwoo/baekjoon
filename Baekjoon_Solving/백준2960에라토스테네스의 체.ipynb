```python
N , K = map(int , input().split())
n=0
a = [False , False] + [True] * (N - 1)
for i in range(2, N+1):
    if a[i]:
        a[i] = False
        n+=1
        if n == K:
            ans = i
            break
        for j in range(2 * i , N+1 , i):
            if a[j]:
                n+=1
                a[j] = False
                if n == K:
                    ans = j
                    break
print(ans)
```
