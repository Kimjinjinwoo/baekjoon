```python
N = int(input())

def sol(N):# 1     Ctrl-A   Ctrl-C  Ctrl-V
    dp = [1,2,3] + [0] * N
    for i in range(3,N):
        dp[i] = dp[i-1] + 1
        for j in range(i - 2):
            dp[i] = max(dp[i] ,dp[j] + dp[j] * (i - j -2) )
    print(dp[N-1])
sol(N)
```
