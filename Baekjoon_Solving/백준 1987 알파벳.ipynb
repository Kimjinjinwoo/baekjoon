```python
## dictionary랑 dfs로 잘하면 되겠는디
import sys
#input = sys.stdin.readline

R, C = map(int, input().split())
board = [input().strip() for _ in range(R)]
visited = [[False] * C for _ in range(R)]
ans = 0

Alphabet_dict = [False] * 26

def dfs(location, length):
    global ans
    ans = max(ans, length)

    x, y = location
    visited[x][y] = True
    alp = board[x][y]
    Alphabet_dict[ord(alp) - ord('A')] = True

    for dx, dy in zip([1,-1,0,0], [0,0,1,-1]):
        nx = x + dx; ny = y + dy
        if 0<= nx < R and 0<= ny < C:
            if not Alphabet_dict[ord(board[nx][ny]) - ord('A')]:
                if not visited[nx][ny]:

                    dfs((nx,ny), length + 1)
    visited[x][y] = False
    Alphabet_dict[ord(alp) - ord('A')] = False

dfs( (0,0), 1)
print(ans)
```
