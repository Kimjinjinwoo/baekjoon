```python
## DP로 각각의 점수 얻을 수 있을거 같은데
### 공백으로 1600점 나오는 개수는 정해져 있음
### 나머지 갯수만 최소로 하면 됨

## 뭘해도 1600보다 점수 낮음
```

```python
n = int(input().strip())
words = []
for _ in range(n):
    words.append(input().strip())
words.sort()
typo = input().strip()
```

```python
import sys
#input = sys.stdin.readline

def char2coord(c):
    char_dict = {}

    row = "QWERTYUIOP"
    coord = (2,2); char_dict[row[0]] = coord
    for r in row[1:]:
        coord = (coord[0] + 4, coord[1])
        char_dict[r] = coord

    row = "ASDFGHJKL"
    coord = (3,6); char_dict[row[0]] = coord
    for r in row[1:]:
        coord = (coord[0] + 4, coord[1])
        char_dict[r] = coord

    row = "ZXCVBNM"
    coord = (5,10); char_dict[row[0]] = coord
    for r in row[1:]:
        coord = ( coord[0] + 4, coord[1] )
        char_dict[r] = coord
    
    return char_dict[c]

def calc_dist(A, B):
    X1 = char2coord(A)
    X2 = char2coord(B)
    return (X1[0] - X2[0]) ** 2 + (X1[1] - X2[1]) ** 2

def get_similarity(A, B):
    A, B = sorted([A, B], key = len) ## A가 길이가 더 작음!
    len_A = len(A); len_B = len(B)
    len_diff = len_B - len_A
    score = len_diff * 1600

    dp = [0] * (len_diff + 1)
    for a_idx in range(len_A):
        new_dp = [0] * (len_diff + 1)
        for ii in range(len_diff + 1):
            new_dp[ii] = min(dp[:ii + 1]) +  calc_dist(A[a_idx], B[a_idx + ii])
        dp = new_dp[:]
        #print(dp)
    return score + min(dp)

def main(n, words, typo):
    score_list = []
    for word in words:
        score_list.append(get_similarity(typo, word))
    
    min_score = min(score_list)

    for score, word in zip(score_list, words):
        if score == min_score:
            print(word)


main(n, words, typo)
```

```python
def char2coord(x : str):
    ToCoord  = {}
    row = "QWERTYUIOP"
    coord = (2,2)
    for character in row:
        ToCoord[character] = coord
        coord = (coord[0] + 4, coord[1])

    row = "ASDFGHJKL"
    coord = (3,6)
    for character in row:
        ToCoord[character] = coord
        coord = (coord[0] + 4, coord[1])

    row = "ZXCVBNM"
    coord = (5,10)
    for character in row:
        ToCoord[character] = coord
        coord = (coord[0] + 4, coord[1])
    return ToCoord[x]

def get_dist(X1, X2):
    if X1 ==' ' or X2 == ' ':
        return 1600
    X1 = char2coord(X1); X2 = char2coord(X2)
    return (X1[0] - X2[0]) ** 2 + (X1[1] - X2[1]) ** 2

def update_dp(A, B, dp, a, ii):
    min_dp = dp[0]; min_idx = 0;
    for jj in range(1, ii+1):
        if dp[jj][1] < min_dp[1]:
            min_dp = dp[jj]
            min_idx = jj

    next_str = min_dp[0] + ' '*(ii - min_idx ) + A[a]
    next_similarity = min_dp[1] + get_dist(A[a], B[a+ii])
    return [next_str, next_similarity]

def get_similarity(A, B):
    len_A = len(A); len_B = len(B); len_diff = abs(len(A) - len(B))
    if( len_A > len_B ):
        A,B = B,A; len_A,len_B = len_B, len_A

    dp = [ ['', 0] for _ in range(len_diff + 1) ]
    for a in range(len_A):
        new_dp = [ ['', 0] for _ in range(len_diff + 1) ]
        for ii in range(len_diff + 1):
            new_dp[ii] = update_dp(A, B, dp, a, ii)
        dp = [new_dp[ii][:] for ii in range(len_diff+ 1)]

    min_str = min(dp, key = lambda x: x[1])[0]
    score = 0
    for x,b in zip(min_str, B):
        score += get_dist(x,b) 
    return score

def main(N : int, words : list, typo : str):
    similarityList = []
    words.sort()

    for word in words:
        similarityList.append(get_similarity(word, typo))
    min_similarity = min(similarityList)   
    for word, similarity in zip(words, similarityList):
        if similarity == min_similarity:
            print(word)

N = int(input())
words = [input().strip() for _ in range(N)]
typo = input().strip()

main(N, words, typo)
```
