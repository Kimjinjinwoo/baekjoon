```python
def sol(A, B):
    length_A = len(A)
    length_B = len(B)
    
    dp = [0] * length_B
    for idxA in range(length_A):
        char_A = A[idxA]
        maxi = 0
        for idxB in range(length_B):
            char_B = B[idxB]
            maxi_ = max(maxi, dp[idxB])
            if (char_A == char_B):
                dp[idxB] = max(dp[idxB], maxi + 1)
            maxi = maxi_
    print(max(dp))
    #print(dp)
```
