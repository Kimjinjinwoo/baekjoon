```python
def sol(N, M, K, animes):
    last_time = M
    how_many_at_once = 0 
    ans = 0

    animes.sort()

    for anime in animes:
        if anime > last_time:
            ans += how_many_at_once
            break
        else:
            how_many_at_once += 1
            if how_many_at_once == K:
                ans += how_many_at_once
                last_time -= anime
                how_many_at_once = 0
    
    print(ans)

N, M, K = map(int, input().split())
animes =list(map(int, input().split()))
sol(N, M, K, animes)
```
