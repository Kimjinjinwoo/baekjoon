```python
# BFS로 안되나..?
def sol(N, HPs):
    cnt = 0    
    while len(HPs) < 3:
        HPs = HPs + [0]

    while sum(HPs):
        HPs.sort()
        HPs.reverse()
        HPs = [max(0, HPs[0] - 9), max(0, HPs[1] - 3), max(0, HPs[2] - 1)]
        #print(HPs)
        cnt += 1
    print(cnt)

N = int(input())
HPs = list(map(int, input().split()))
sol(N, HPs)
```
