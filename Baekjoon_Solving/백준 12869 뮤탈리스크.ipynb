```python
from collections import deque

def sol(N, HPs):
    while len(HPs) < 3:
        HPs = HPs + [0]
    x,y,z = HPs
    dp = [ [[0]*(z+1) for _ in range(y+1)] for _ in range(x+1) ]
    dp[x][y][z] = 1
    for i in range(x, -1, -1):
        for j in range(y, -1, -1):
            for k in range(z, -1, -1):
                if not dp[i][j][k]:
                    continue
                for order in [(0,1,2), (0,2,1), (1,0,2), (1,2,0), (2,0,1), (2,1,0)]:
                    damage = [9,3,1]
                    damage_x, damage_y, damage_z = damage[order[0]], damage[order[1]], damage[order[2]]
                    x_, y_, z_ =  max(i - damage_x, 0), max(j - damage_y, 0), max(k - damage_z, 0)
                    #print(i,j,k)
                    if dp[x_][y_][z_] == 0:
                        dp[x_][y_][z_] = dp[i][j][k] + 1 
                    dp[x_][y_][z_] = min(dp[i][j][k]+1, dp[x_][y_][z_])

    print(dp[0][0][0]-1)

N = int(input())
HPs = list(map(int, input().split()))
sol(N, HPs)
```
