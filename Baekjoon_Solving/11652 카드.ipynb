```python
t = int(input())
nums = {}
max = 0
idx = 0

for _ in range(t):
    num =int(input())
    if num not in nums:
        val = 1
        nums[num] = val
    else:
        nums[num] += 1
        val = nums[num]

    if val > max:
        max = val
        idx = num
    elif val == max:
        idx = min(idx , num)

print(idx)
```
