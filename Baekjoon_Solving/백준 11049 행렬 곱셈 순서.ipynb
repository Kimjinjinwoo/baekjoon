```python
from sys import maxsize
import sys
input = sys.stdin.readline

def sol(N, MatrixList):
    dp = [[0] * N for _ in range(N)]

    for size in range(1,N): 
        for start in range(N-1):
            end = start + size
            if end >= N:
                break
            next = maxsize
            for mid in range(start, end): ## start ~ end - 1  
                next = min(next, dp[start][mid] + dp[mid+1][end] + MatrixList[start][0] * MatrixList[mid][1] * MatrixList[end][1])
            dp[start][end] = next
    print(dp[0][-1])
    
N = int(input())
MatrixList = [tuple(map(int, input().split())) for _ in range(N)]

sol(N, MatrixList)
```

```python
## 재귀 + 메모이제이션으로 풀어 봐야지~
from sys import maxsize
import sys
#input = sys.stdin.readline

N = int(input())
MatrixList = [tuple(map(int, input().split())) for _ in range(N)]
memo = [[-1] * N for _ in range(N)]

def recursion(start, end):
    if memo[start][end] != -1:
        return memo[start][end]
    if start == end:
        memo[start][end] = 0
        return 0
    else:
        next = maxsize
        for mid in range(start, end):
            tmp = recursion(start, mid) + recursion(mid+1, end) + MatrixList[start][0] * MatrixList[mid][1] * MatrixList[end][1]
            next = min(tmp, next)
        memo[start][end] = next
        return next

print(recursion(0,N-1))
```
