```python
import sys
input = sys.stdin.readline

N , M = map(int , input().split())
scores = list(map(int , input().split()))

def score_mapping(scoring):
    Id = int(scoring[0])
    total = 0
    for O_X , score in zip(scoring[1:] , scores):
        if O_X == 'O':
            total += score
    return [-1 * total , Id]

X = []

for _ in range(M):
    xxx = score_mapping(input().split())
    X.append(xxx)

X.sort()
print(X[0][1] ,  -1 * X[0][0])
```
