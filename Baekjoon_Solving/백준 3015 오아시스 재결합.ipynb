```python
from collections import deque
import sys

def get_input():
    input = sys.stdin.readline
    N = int(input())
    height_list = [int(input()) for _ in range(N)]

    return N, height_list

def get_cnt_dict(nums):
    cnt_dict = {}
    for num in list(set(nums)):
        cnt_dict[num] = 0
    return cnt_dict

def solve(N, height_list):
    stack = []
    stack_len = 0
    ans = 0

    for height in height_list:
        #print(ans, cnt_dict)
        while(stack):
            top = stack.pop()

            if top < height:
                ans += cnt_dict[top]
                cnt_dict[top] = 0
                continue
            ## 같은 경우에 대한 예외 처리
            elif top == height:
                ans += cnt_dict[top]
                #stack.append(top)
                #print(stack, cnt_dict[top], top)
                if stack:
                    ans += 1
                break
            else:
                ans += 1
                stack.append(top)
                break
        stack.append(height)
        cnt_dict[height] += 1

    print(ans)

## 예시 입력 : 2 4 1 2 2 5 1
## 예시 정답 : 10
## 풀이 :   (2, 4), (4, 1), (4, 2), (4, 2), (4, 5)
##          (1, 2), (2, 2), (2, 5), (2, 5), (5, 1)
## 일단 높은 거 나오면 stack 초기화 될 거임... (이건 맞나?)

if __name__ == "__main__":
    N, height_list = get_input()
    cnt_dict = get_cnt_dict(height_list)
    solve(N, height_list)
```
