## A번 On My Way Dorm

```python
N, S =map(int , input().split())
performanceList = list(map(int,input().split()))
command = input().strip()
```

```python
from collections import deque
import sys
#input = sys.stdin.readline

def sol(N, S, performanceList, command):
    floor = calc_floor(S, performanceList, command)
    ans = how_to_leave(N, S, performanceList, floor)
    print(ans)
    return

def get_Command_to_Sign():
    Command_to_Sign = {}
    Command_to_Sign['+'] = 1
    Command_to_Sign['-'] = -1
    Command_to_Sign['0'] = 0
    return Command_to_Sign

def calc_floor(S, performanceList, command):
    now = S
    velocity = 0
    Command_to_Sign = get_Command_to_Sign()
    for comm in command:
        velocity += Command_to_Sign[comm] * performanceList[now-1]
        now += velocity
    return now

def how_to_leave(N, S, performanceList, now):
    # bfs로 풀면 될라나
    # (위치, velocity, command)
    visited = [[False] * (2*N + 2) for _ in range(N+1)] ## (층, 속도)
    Command_to_Sign = get_Command_to_Sign()
    start = (now, 0, '')
    visited[now][0] = True
    queue = deque([start])
    
    while queue:
        now, velocity, command = queue.popleft()
        for comm in "-0+":
            next_velocity = velocity + Command_to_Sign[comm] * performanceList[now-1]
            next_command = command + comm
            next = now + next_velocity
            if 0 < next < N+1:
                #print((next, next_velocity, next_command))
                if next == S and next_velocity == 0:
                    return next_command
                if next_velocity != 0:
                    if not visited[next][next_velocity]:
                        visited[next][next_velocity] = True
                        queue.append( (next, next_velocity, next_command) )
    return

#N, S =map(int , input().split())
#performanceList = list(map(int,input().split()))
#command = input().strip()

sol(N, S, performanceList, command)
```
