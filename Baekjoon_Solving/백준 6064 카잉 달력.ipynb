```python
import sys
input = sys.stdin.readline

def sol():
    T = int(input())
    for _ in range(T):
        M,N,x,y = map(int, input().split())
        find_year( M, N, x, y )

## 최소 공배수가 마지막 해임~
def gcd(A,B):
    A , B = max(A,B) , min(A,B)
    while B>0:
        A, B = B , A % B
    return A

def lcm(A,B):
    return A * B // gcd(A,B)

def find_year(M,N,x,y):
    ans = -1
    last_year = lcm(M,N)
    remainder_x = (x-1) % M 
    for n in range(remainder_x, last_year, M):
        if n % N == y - 1:
            ans = n + 1
    print(ans)

sol()
```
