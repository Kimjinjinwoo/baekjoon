```python
N = int(input())

tree = [0 for _ in range(N)]

pre_ans = ''
in_ans = ''
post_ans = ''


def to_number(A):
    return ord(A) - ord('A')

def to_str(n):
    return chr(n + ord('A'))

dot = to_number('.')

def preorder(parent = 0):
    global pre_ans
    left , right = tree[parent]
    pre_ans += to_str(parent)
    #print(parent , pre_ans)
    if left != dot:
        preorder(left)
    if right != dot:
        preorder(right)

def inorder(parent = 0):
    global in_ans
    left , right = tree[parent]
    if left != dot:
        inorder(left)
    in_ans += to_str(parent)
    if right != dot:
        inorder(right)

def postorder(parent = 0):
    global post_ans
    left , right = tree[parent]
    if left != dot:
        postorder(left)
    if right != dot:
        postorder(right)
    post_ans += to_str(parent)

for n in range(N):
    parent , left , right = map(to_number , input().split()) 
    tree[parent]=((left , right))

preorder()
inorder()
postorder()

print(pre_ans)
print(in_ans)
print(post_ans)
```
