```python
### 트리 구조 알아내는 게 쉽지 않겠네 ㄷㄷ

import sys
sys.setrecursionlimit(10** 6)

def sol(N, links):
    tree = [[] for _ in range(N+1)]
    node_info = [[0,0]] * (N+1)
    for link in links:
        X = link.pop(0)
        while True:
            y = link.pop(0)
            if y != -1:
                w = link.pop(0)
                tree[X].append( (y, w) )
            else:
                break
    node_info = node_info_dfs(1, tree, node_info)

    maxi = 0
    for i in range(1,N+1):
        node_info[i].sort()
        if node_info[i][-1] + node_info[i][-2] > maxi:
            maxi = node_info[i][-1] + node_info[i][-2]
    print(maxi)

def node_info_dfs(now, tree, node_info):
    if not tree[now]: ## 자식 노드가 비어 있다면
        node_info[now] = [0,0]
    else: ##자식 노드가 있으면
        node = [0] * max(2, len(tree[now]))
        for i, son in enumerate(tree[now]):
            y, w = son
            node_info = node_info_dfs(y, tree, node_info)
            node[i] = max(node_info[y]) + w
        node_info[now] = node
    return node_info

N = int(input())
links = [list(map(int,input().split())) for _ in range(N)]
sol(N, links)
```
