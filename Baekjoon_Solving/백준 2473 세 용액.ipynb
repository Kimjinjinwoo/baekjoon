```python
import sys
input = sys.stdin.readline

N = int(input())
nums = sorted(list(map(int, input().split())))

def BiSearch(value, start, end):
    ## value와 가장 가까운 수 찾기!
    ans = (end + start) // 2
    while (start <= end):
        mid = (end + start) // 2
        num = nums[mid]

        if abs(num + value) < abs(nums[ans] + value):
            ans = mid

        if num + value < 0:
            start = mid + 1
        elif num + value > 0:
            end = mid - 1
        elif num + value== 0:
            break

    return ans

k = 0
start, end = 1, N-1

min_value = nums[k] + nums[start] + nums[end]
ans = [0, 1, N-1]

for k in range(N):
    num_k = nums[k]
    start, end = 0, N-1
    if start == k:
        start += 1
    if end == k:
        end -= 1

    while (start < end):
        value = num_k + nums[start] + nums[end]

        if abs(value) < abs(min_value):
            min_value = value
            ans = [num_k, nums[start], nums[end]]

        if value > 0:
            end -= 1
            if end == k:
                end -= 1
        elif value < 0:
            start += 1
            if start == k:
                start += 1
        elif value == 0:
            break

ans.sort()
print(*ans)
```
