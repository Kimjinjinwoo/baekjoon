```python
N = int(input())

board_ = [list(map(int,  input().split())) for _ in range(N)]
white = 0; blue = 0

#print(board_)

def quarter(board , N):
    
    n = N // 2
    board1 = board[:n]
    board1 = [ k[:n] for k in board1 ]
    board2 = board[n:]
    board2 = [ k[:n] for k in board2 ]

    board3 = board[:n]
    board3 = [ k[n:] for k in board3 ]
    board4 = board[n:]
    board4 = [ k[n:] for k in board4 ]

    return [board1 , board2 , board3 , board4]
now_board = []
now_N = 0

def gogo(board  , N):
    global white , blue , now_N
    if N == 1:
        if board[0][0]:
            white += 1
        else:
            blue += 1
        return
    unity = True
    color = board[0][0]


    for i in range(N):
        for j in range(N):
            if color != board[i][j]:
                unity = False
                break
    
    if unity:
        if color:
            white += 1
        else:
            blue += 1

    else:
        board_2 = [n[:]  for n in board]
        four_ = quarter(board_2 , N)
        for now in four_:
            now_ = [n[:]  for n in now]
            gogo( now_ , N // 2)

gogo(board_ , N)
#색깔 반대로 적었음
print(blue)
print(white)
```
