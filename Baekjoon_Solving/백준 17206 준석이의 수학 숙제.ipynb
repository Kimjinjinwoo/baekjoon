```python
T = int(input())
nums = list(map(int , input().split()))

def sub_calcul(n , d):
    maxi = (n//d) * d 
    result = 0
    if maxi >= d :
        result = (maxi + d) * (n//d) // 2
    return result

def calcul(n):
    result = sub_calcul(n , 3) + sub_calcul(n , 7) - sub_calcul(n , 21)
    return result

for n in nums:
    print(calcul(n))
```
