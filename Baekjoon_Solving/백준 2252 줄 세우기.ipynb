```python
from collections import deque
import sys
input = sys.stdin.readline

def sol(N, M, compList):
    graph = make_graph(N, M, compList)
    adjacency = make_adjacency(N, graph)
    result = topological_sort(N,graph, adjacency)
    print(' '.join(map(str, result)))
    return

def make_graph(N, M, compList):
    graph = [[] for _ in range(N+1)]
    for small, big in compList:
        graph[small].append(big)
    return graph

def make_adjacency(N, graph):
    adjacency = [0] * (N+1)
    for linkeds in graph[1:]:
        for next in linkeds:
            adjacency[next] += 1
    return adjacency

def topological_sort(N,graph, adjacency):
    visited = [False] * (N+1)
    result = []
    queue = deque()
    queue, visited = Init_queue(queue, adjacency, visited)
    while queue:
        #print(result)
        x = queue.popleft()
        result.append(x)
        adjacency, queue, visited = node_one_step(x,graph, adjacency, queue, visited)
        
    return result

def Init_queue(queue, adjacency, visited):
    for ii in range(1,N+1):
        if adjacency[ii] == 0:
            if not visited[ii]:
                queue.append(ii)
                visited[ii] = True
    return queue, visited

def node_one_step(x, graph, adjacency, queue, visited):
    for next in graph[x]:
        adjacency[next] -= 1
        if adjacency[next] == 0:
            if not visited[next]:
                queue.append(next)
                visited[next] = True
    return adjacency, queue, visited


N, M = map(int, input().split())
compList = [tuple(map(int, input().split())) for _ in range(M)]
sol(N, M, compList)
```
