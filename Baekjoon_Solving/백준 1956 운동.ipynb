```python
## 플로이드-워셜로 최단거리 다 찾아 놓고
## 두점 길이 최소 인거 찾기
## 또는 플로이드-워셜 애초에 사이클 되나 확인
from sys import maxsize as inf
import sys

def get_input():
    input = sys.stdin.readline
    V, E = map(int, input().split())
    graph = [tuple(map(int, input().split())) for _ in range(E)] ## (a, b, cost)
    return V, E, graph

def init_dist(V, E, graph):
    dist_graph = {ii : {jj : inf for jj in range(1,V+1)} for ii in range(1,V+1)}
    for a, b, cost in graph:
        dist_graph[a][b] = min(cost, dist_graph[a][b]) ## a->b 의 거리가 cost


    return dist_graph

def Floyd_Warshall(V, E, graph):
    dist_graph = init_dist(V, E, graph)
    #dist_graph = [row[:] for row in initial_dist_graph] ## copy하기

    for mid in range(1, V+1):
        for start in range(1, V+1):
            for end in range(1, V+1):
                dist_graph[start][end] = min(dist_graph[start][end],
                                             dist_graph[start][mid] + dist_graph[mid][end])
    return dist_graph

def get_ans(V, dist_graph):
    ans = inf
    for ii in range(1, V+1):
        ans = min(ans, dist_graph[ii][ii])
    if ans == inf:
        return -1
    else:
        return ans

if __name__ == "__main__":
    V, E, graph = get_input()

    dist_graph = Floyd_Warshall(V, E, graph)

    ans = get_ans(V, dist_graph)
    print(ans)
```
