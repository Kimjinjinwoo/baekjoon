```python
N,M = map(int ,input().split())
maze = [list(map(int, input().split())) for _ in range(N)]
dp = [ [0]*M for _ in range(N) ]

def go(i,j):
    can_go_i = [i ,i+1 , j+1]
    can_go_j = [j+1 , j , j+1]

    for ii, jj in zip(can_go_i , can_go_j):
        if 0<=ii<N and 0<=jj<M:
            dp[ii][jj] = max(dp[i][j] + maze[i][j] , dp[ii][jj])

for i in range(N):
    for j in range(M):
        go(i,j)

print(dp[N-1][M-1] + maze[N-1][M-1])
```
