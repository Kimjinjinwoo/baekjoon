```python
## 브루트포스로 풀자~~
import sys
from collections import deque
input = sys.stdin.readline

def sol(N, M, graph):
    answer = 0
    zeros, number_of_zeros = find_zeros(N, M, graph)
    for i in range(number_of_zeros-2):
        x1,y1 = zeros[i]
        graph[x1][y1] = 1
        for j in range(i+1, number_of_zeros-1):
            x2,y2 = zeros[j]
            graph[x2][y2] = 1
            for k in range(j+1, number_of_zeros):
                x3,y3 = zeros[k]
                graph[x3][y3] = 1
                copied_graph = [row[:] for row in graph]
                safty = infection(N, M, copied_graph)
                answer = max(answer, safty)
                graph[x3][y3] = 0
            graph[x2][y2] = 0
        graph[x1][y1] = 0
    print(answer)
    return

def infection(N, M, graph):
    visited = [[False]*M for _ in range(N)]
    for i in range(N):
        for j in range(M):
            if visited[i][j]:
                continue
            if graph[i][j] == 2:
                graph, visited = infection_bfs(N, M, i, j, graph, visited)
    safty = calc_safty(N, M, graph)
    return safty

def calc_safty(N, M, graph):
    safty = 0
    for i in range(N):
        for j in range(M):
            if graph[i][j] == 0:
                safty += 1
    return safty

def infection_bfs(N, M, i, j, graph, visited):
    queue = deque([(i, j)])
    while queue:
        x, y = queue.pop()
        nx = [x+1, x-1, x, x]
        ny = [y, y, y+1, y-1]
        for x, y in zip(nx, ny):
            if 0<=x<N and 0<=y<M:
                if graph[x][y] == 0:
                    graph[x][y] = 2
                    visited[x][y] = True
                    queue.append((x, y))
    return graph, visited

def find_zeros(N, M, graph):
    zeros = []
    number_of_zero = 0
    for i in range(N):
        for j in range(M):
            if graph[i][j] == 0:
                zeros.append((i,j))
                number_of_zero += 1
    return zeros, number_of_zero

N, M = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(N)]
sol(N, M, graph)
```
