```python
N, M = map(int, input().split())
nums = list(map(int, input().split()))

start, end = 0, 1
ans = 0
acc_sum = [0]
for num in nums:
    acc_sum.append(acc_sum[-1] + num)

while (end <= N):
    value = acc_sum[end] - acc_sum[start]

    if value == M:
        ans += 1

    if value <= M:
        end    +=   1
    elif value > M:
        start  +=   1

    if start == end:
        end += 1

print(ans)
```
