```python
from sys import maxsize
import sys

input = sys.stdin.readline

N,M = map(int , input().split())

space = [list(map(int, input().split())) for _ in range(N)]

def go(now , next , M = M):
    result = [[maxsize]*3 for _ in range(M)]

    for n in range(M):  # 출발할 이전 위치
        for nd in [-1,0,1]: #이동할려는 방향
            for ld in range(3):
                if nd == ld - 1 :   # 두 번 연속 같은지 확인 
                    continue
                to_go = n + nd      # 이제 어디로 가는 지
                if 0<= to_go <M:    # 범위 내에 있는 지
                    if result[to_go][nd + 1] > now[n][ld] + next[to_go]:
                        result[to_go][nd + 1] = now[n][ld] + next[to_go]
    return result

now = [[space[0][i]] * 3 for i in range(M)] 

for i in range(1,N):
    now = go(now , space[i])
    #print(now)
print(min(map(min , now)))
```

```python
from sys import maxsize
import sys

N,M = map(int , input().split())

space = [list(map(int, input().split())) for _ in range(N)]

def go(now , next , last_move , w = M):
    result = [maxsize] * M
    now_move = [[]] * M
    for n in range(M):
        lms = last_move[n]
        for lm in lms:
            for nm in [-1,0,1]:
                if nm == lm:
                    continue
                nd = n + nm
                if 0 <= nd < M:
                    fuel = now[n] + next[nd]

                    if fuel < result[nd]: 
                        result[nd] = fuel
                        now_move[nd] = [nm]
                    elif fuel == result[nd]:
                        result[nd] = fuel
                        now_move[nd].append(nm)
    return result , now_move

now = space[0]; last_move = [[3]] * M

for i in range(1,N):
    now , last_move  = go(now , space[i] , last_move)
    #print(now)
print(min(now))
```
