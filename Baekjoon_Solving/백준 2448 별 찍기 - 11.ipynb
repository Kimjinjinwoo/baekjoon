```python
w_dict = {3 : 5}
h_dict = {3 : 3}

for k in range(1,11):
    N = 3 * (2 ** k)
    w_dict[N] = w_dict[3 * (2 ** (k-1))] * 2 + 1
    h_dict[N] = N

N = int(input())

whole = [[' '] * w_dict[N] for _ in range(h_dict[N])]

def tri(size, bottom, top, left, right):
    size = top - bottom + 1
    if size == 3:
        for ii in range(left, right+1):
            whole[bottom][ii] = '*'
        whole[bottom+1][left+1] = '*'
        whole[bottom+1][right-1] = '*'
        whole[bottom+2][left+2] = '*'
        return
    #print(size)
    before_size = size // 2
    before_w = w_dict[before_size]
    before_h = h_dict[before_size]

    middle_point = left + (right - left) // 2

    ##tri_top
    top_top = top
    top_bottom = bottom + before_h
    top_left = middle_point - before_w // 2
    top_right = middle_point + before_w // 2
    tri(before_size, top_bottom, top_top, top_left, top_right)

    ##tri_left
    left_top = bottom + before_h - 1
    left_bottom = bottom
    left_left = left
    left_right = middle_point - 1
    tri(before_size, left_bottom, left_top, left_left, left_right)

    ##tri_right
    right_top = bottom + before_h - 1
    right_bottom = bottom
    right_left = middle_point + 1
    right_right = right
    tri(before_size, right_bottom, right_top, right_left, right_right)

tri(N, 0, h_dict[N] - 1, 0, w_dict[N] - 1)

for ii in range(h_dict[N])[::-1]:
    print(''.join(whole[ii]))
```
