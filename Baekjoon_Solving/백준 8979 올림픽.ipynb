```python
import sys
input = sys.stdin.readline

N , M = map(int , input().split())
countrys = []
for _ in range(N):
    country , G , S , B = map(int, input().split())
    if country == M:
        country = -1
    countrys.append((-G,-S,-B,country))
countrys.sort()

for i in range(N):
    if countrys[i][3] == -1:
        print(i+1)
        break
```
