```python
def sol(A, B):
    length_A = len(A)
    length_B = len(B)
    ## 그냥 편의상
    if length_B < length_A:
        tmp = A; A = B; B = tmp;
        length_A = len(A)
        length_B = len(B)

    dp = [0] * length_B
    maxi = 0
    for idxA in range(length_A):
        dp_2 = [0] * length_B
        for idxB in range(length_B):
            x = A[idxA]; y = B[idxB]
            if x==y:
                if idxB == 0:
                    cnt = 1
                else:
                    cnt = dp[idxB-1] + 1
                dp_2[idxB] = cnt
        maxi = max(maxi, max(dp_2))
        dp = dp_2
        
    print(maxi)
```
