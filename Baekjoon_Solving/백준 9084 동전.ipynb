```python
import sys
#input = sys.stdin.readline

def sol(N, M, coinList):
    dp = [0] * (M+1)
    for ii, coin in enumerate(coinList):
        if coin > M:
            break
        dp[coin] += 1
        for ii in range(0, M-coin+1):
            if dp[ii]:
                dp[ii+ coin] += dp[ii]
    print(dp[M])
    return

T = int(input())

for _ in range(T):
    N = int(input())
    coinList = list(map(int, input().split()))
    M = int(input())
    sol(N, M, coinList)
```
