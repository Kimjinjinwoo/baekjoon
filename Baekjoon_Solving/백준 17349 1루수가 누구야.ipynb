```python
import sys
input = sys.stdin.readline

def sol():
    guess = []
    tellings = [ list(map(int , input().split())) for _ in range(9) ]

    for n in range(9):
        Y, N = n_is_lier(tellings , n)
        first_baseman = verify_who(Y,N, guess)
        #print(n)
        #print(Y);print(N);print(first_baseman);print()
        if first_baseman != -1:
            guess.append(first_baseman) 
    
    guess = list(set(guess))
    if len(guess) == 1:
        print(guess[0])
    else:
        print(-1)

def n_is_lier(tellings , n):
    tells = tellings[:]
    yes = [0] * 10
    no = [0] * 10
    for i in range(9):
        yn , name = tells[i]
        if i == n:
            yn = 1 - yn
        if yn == 1:
            yes[name] = 1
        else:
            no[name] = 1
    return yes, no


def verify_who(Y , N , guess):
    yes = Y[:]; no = N[:]
    ans = -1
    contradict = False
    for i in range(1,10):
        if yes[i] == 1:
            if no[i] == 1:
                contradict = True
            elif ans != -1:
                contradict = True
            else:
                ans = i

    if sum(yes) == 0:
        for i in range(1,10):
            if no[i] == 0:
                guess.append(i)
    if contradict:
        ans = -1

    return ans
                
sol()
```
