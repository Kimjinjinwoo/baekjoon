```python
import sys
input = sys.stdin.readline

def sol():
    ans = []
    tellings = [ list(map(int , input().split())) for _ in range(9) ]

    for n in range(9):
        Y, N = n_is_lier(tellings , n)
        first_baseman = verify_who(Y,N)
        if first_baseman != -1:
            ans.append(first_baseman) 
    
    ans = list(set(ans))
    if len(ans) == 1:
        print(ans[0])
    else:
        print(-1)

def n_is_lier(tells , n):
    yes = [0] * 10
    no = [0] * 10
    for i in range(9):
        yn , name = tells[i]
        if i == n:
            yn = 1 - yn
        if yn == 1:
            yes[name] = 1
        else:
            no[name] = 1
    return yes, no


def verify_who(yes , no):
    ans = -1
    contradict = False
    for i in range(1,10):
        if yes[i] == 1:
            if no[i] == 1:
                contradict = True
            elif ans != -1:
                contradict = True
            else:
                ans = i

    if sum(yes) == 0:
        if sum(no) == 8:
            for i in range(1,10):
                if no[i] == 0:
                    ans = i
    if contradict:
        ans = -1

    return ans
                
sol()
```
