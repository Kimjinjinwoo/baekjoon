```python
import sys
#input = sys.stdin.readline

def sol():
    N = int(int(input()))
    room = [list(map(int, input().split())) for _ in range(N)]
    print(verify(room, N))

def verify(room, N):
    professor, seong = find_professor_and_seong(room, N)
    r = dist(professor, seong)
    if r < 5:
        return 0
    else:
        people = count_people(room, professor, seong, N)
        if people >= 3:
            return 1
        else:
            return 0

def find_professor_and_seong(room, N):
    for y in range(N):
        for x in range(N):
            if room[x][y] == 5:
                professor = (x, y)
            elif room[x][y] == 2:
                seong = (x, y)
    return professor, seong

def dist(A, B):
    return ( (A[0] - B[0])**2 + (A[1] - B[1])**2 )**(1/2)
        
def count_people(room, professor, seong, N):
    cnt = 0
    px,py = professor
    sx,sy = seong
    px,sx = min(px,sx) , max(px,sx) # ㅋㅋ 코드 더러운데 귀찮어
    py,sy = min(py,sy) , max(py,sy)
    for y in range(py, sy+1):
        for x in range(px, sx+1):
            if room[x][y] == 1:
                cnt+=1
    return cnt

sol()
```
