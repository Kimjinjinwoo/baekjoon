```python
import sys
import heapq

N = int(input())
M = int(input())
graph = [[] for _ in range(N)]
visited = [False] * N

for _ in range(M):
    v1,v2,w = map(int, sys.stdin.readline().split())
    v1 -= 1 ; v2 -= 1
    graph[v1].append([w,v1,v2])
    graph[v2].append([w,v2,v1])

def prim():
    S = 0
    candi = graph[0]    # 0에서 시작하기
    visited[0] = True

    heapq.heapify(candi)

    while candi:
        # 일단 가능한 최솟값 찾기
        w,v1,v2 = heapq.heappop(candi)
        if not visited[v2]:
            S += w
            visited[v2] = True
        
            for w_,v1_,v2_ in graph[v2]:
                if not visited[v2_]:
                    heapq.heappush(candi , [w_,v1_,v2_])
    return S

print(prim())

```
