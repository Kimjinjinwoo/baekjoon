```python
import sys

sys.setrecursionlimit(10**6)

V , E = map(int , sys.stdin.readline().split() )
visited = [ False ] * V
visited[0] = True
graph = [ [] for _ in range(V)]

for _ in range(E):
    v1 , v2 , w = map(int , sys.stdin.readline().split())
    v1 -= 1; v2 -= 1
    # 만약 이미 선이 있으면
    graph[v1].append([v2,w])
    graph[v2].append([v1,w])


S = 0

def find_next():
    global S; global V
    mini_v1=None ; mini_v2 = None ; mini_w = 2147483648
    for v1 in range(V):
        if visited[v1]: # 방문 했던 점이면
            for info_ in graph[v1]:
                v2 , w = info_
                if w == 0:
                    continue
                if visited[v2]:
                    continue
                #print(w)
                if w < mini_w:
                    mini_v1 = v1
                    mini_v2 = v2
                    mini_w = w
    if mini_v1 == None:
        return
    else:
        #print(mini_v1 , mini_v2 , mini_w)
        visited[mini_v1] = True
        visited[mini_v2] = True
        S += mini_w
        find_next()
    
while sum(visited) <V:
    find_next()

print(S)
```

```python
import sys
import heapq

sys.setrecursionlimit(10**6)

V , E = map(int , input().split() )
visited = [ False ] * V
visited[0] = True
graph = [ [] for _ in range(V)]

for _ in range(E):
    v1 , v2 , w = map(int , sys.stdin.readline().split())
    v1 -= 1; v2 -= 1
    # 만약 이미 선이 있으면
    graph[v1].append([w,v1,v2])
    graph[v2].append([w,v2,v1])

S = 0

def prim(graph , start_node):
    visited[start_node] = True
    candidate = graph[start_node]
    heapq.heapify(candidate)
    mst = []
    total_weight = 0

    while candidate:
        weight , v1 , v2 = heapq.heappop(candidate)
        if not visited[v2]:
            visited[v2] = True
            mst.append((v1 , v2))
            total_weight += weight 

            for edge in graph[v2]:
                if not visited[edge[2]]:
                    heapq.heappush(candidate , edge)
    return total_weight

    

print(prim(graph , 0))
```

```python
V , E = map(int , input().split() )
visited = [ False ] * V
graph = [ [0 for _ in range(V)] for _ in range(V)]
edges = [ list(map( int , input().split())) for _ in range(E) ]
S = 0

def find_next():
    global S;
    mini_v1=None ; mini_v2 = None ; mini_w = 2147483648
    for v1 , visit_q in enumerate(visited):
        if visit_q: # 방문 했던 점이면
            for v2,w in enumerate(graph[v1]):
                if w == 0 or visited[v2]:
                    continue
                if w < mini_w:
                    mini_v1 = v1
                    mini_v2 = v2
                    mini_w = w
        if mini_v1 == None:
            return
        else:
            #print(mini_v1 , mini_v2 , mini_w)
            visited[mini_v2] = True
            graph[mini_v1][mini_v2] = 0
            graph[mini_v2][mini_v1] = 0
            S += mini_w
            find_next()
                      
            
for v1 , v2 , w in edges:
    v1 -= 1; v2 -= 1
    # 만약 이미 선이 있으면
    if graph[v1][v2]:
        # 가중치 작은 거 고르기
        if w > graph[v1][v2]:
            break
    graph[v1][v2] = w
    graph[v2][v1] = w

visited[0] = True

find_next()
print(S)
```
